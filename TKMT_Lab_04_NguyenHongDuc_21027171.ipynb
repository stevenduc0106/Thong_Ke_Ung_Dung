**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import random
from scipy.stats import binom, norm

def baiToanSimulasiSinhVien():
    soCauDung = 0
    for _ in range(10):
        luaChon = random.randint(1, 4)  # Chọn ngẫu nhiên một phương án từ 1 đến 4
        if luaChon == 1:  # Giả sử phương án 1 là đáp án đúng
            soCauDung += 1
    return soCauDung

# a. In ra số câu đúng của sinh viên trong 10 lần kiểm tra
for _ in range(10):
    soCauDung = baiToanSimulasiSinhVien()
    print(f"Số câu đúng: {soCauDung}")

# b. Tính xác suất để sinh viên được 4 điểm
soKetQuaThanhCong = binom.pmf(4, 10, 1/4)  # Sử dụng phân phối nhị thức (binomial distribution)
print(f"Xác suất đạt 4 điểm: {soKetQuaThanhCong}")

# c. Tính Kỳ vọng (E(X)) và Phương sai (Var(X))
trungBinh = 10 * (1/4)  # E(X) = n * p
phuongSai = 10 * (1/4) * (3/4)  # Var(X) = n * p * (1 - p)

print(f"Kỳ vọng (E(X)): {trungBinh}")
print(f"Phương sai (Var(X)): {phuongSai}")
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
from scipy.stats import poisson

lambda_value = 3  # Số cuộc gọi mỗi phút

# Tính xác suất nhận được 1 cuộc gọi trong vòng một phút
prob_1_call = poisson.pmf(1, lambda_value)
print(f"Xác suất nhận được 1 cuộc gọi: {prob_1_call}")

# Tính xác suất nhận được 2 cuộc gọi trong vòng một phút
prob_2_calls = poisson.pmf(2, lambda_value)
print(f"Xác suất nhận được 2 cuộc gọi: {prob_2_calls}")

# Tính xác suất nhận được 3 cuộc gọi trong vòng một phút
prob_3_calls = poisson.pmf(3, lambda_value)
print(f"Xác suất nhận được 3 cuộc gọi: {prob_3_calls}")
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.
b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import binom, norm

# a. Tính tỷ lệ của mỗi loại trái cây
mu = 500  # Trung bình
sigma = np.sqrt(16)  # Độ lệch chuẩn

prob_type1 = 1 - norm.cdf(505, mu, sigma)  # Xác suất trên 505 gam
prob_type2 = norm.cdf(505, mu, sigma) - norm.cdf(495, mu, sigma)  # Xác suất từ 495 - 505 gam
prob_type3 = norm.cdf(495, mu, sigma)  # Xác suất dưới 495 gam

print(f"Tỷ lệ Loại 1: {prob_type1}")
print(f"Tỷ lệ Loại 2: {prob_type2}")
print(f"Tỷ lệ Loại 3: {prob_type3}")

# b. Mô phỏng lấy mẫu 20 trái cây
sample_size = 20
sample = np.random.normal(mu, sigma, sample_size)

# Vẽ histogram
plt.hist(sample, bins=10, alpha=0.5)
plt.xlabel("Trọng lượng (gam)")
plt.ylabel("Số lượng")
plt.title("Histogram trọng lượng trái cây (mẫu 20)")
plt.show()

# Tính giá trị trung bình và độ lệch chuẩn từ mẫu
mean_sample = np.mean(sample)
std_sample = np.std(sample)

print(f"Trung bình từ mẫu 20: {mean_sample}")
print(f"Độ lệch chuẩn từ mẫu 20: {std_sample}")

# c. Mô phỏng lấy mẫu 100 trái cây
sample_size = 100
sample = np.random.normal(mu, sigma, sample_size)

# Vẽ histogram
plt.hist(sample, bins=10, alpha=0.5)
plt.xlabel("Trọng lượng (gam)")
plt.ylabel("Số lượng")
plt.title("Histogram trọng lượng trái cây (mẫu 100)")
plt.show()

# Tính giá trị trung bình và độ lệch chuẩn từ mẫu
mean_sample = np.mean(sample)
std_sample = np.std(sample)

print(f"Trung bình từ mẫu 100: {mean_sample}")
print(f"Độ lệch chuẩn từ mẫu 100: {std_sample}")

# d. Nhận xét
print("Nhận xét:")
print("Càng lấy mẫu càng lớn (100 trái cây), giá trị trung bình và độ lệch chuẩn của trọng lượng trái cây từ mẫu càng xấp xỉ giá trị trung bình và độ lệch chuẩn của phân phối chuẩn ban đầu (500 gam và độ lệch chuẩn 4 gam). Điều này phản ánh định luật số lớn của định lý giới hạn trung tâm (Central Limit Theorem), cho thấy khi kích thước mẫu càng lớn, phân phối của trung bình mẫu sẽ xấp xỉ phân phối chuẩn.")
```
