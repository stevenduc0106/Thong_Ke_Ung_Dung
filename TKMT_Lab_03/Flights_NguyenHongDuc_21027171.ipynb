```python
import pandas as pd

df = pd.read_csv('flights.csv')
df.head()
```

```python
#Xoá cột Unnamed: 0
df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Tạo cột mới 'year_month' kết hợp 'year' và 'month'
df['year_month'] = df['year'].astype(str) + '-' + df['month']

# Nhóm dữ liệu theo 'year_month' và tính tổng số lượng hành khách
passengers_per_month = df.groupby('year_month')['passengers'].sum().reset_index()

# Xem dữ liệu đã lọc và tính tổng
print(passengers_per_month)
```
