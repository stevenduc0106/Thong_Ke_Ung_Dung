```python
# Cau 1: Nhận thức được rằng phần thưởng cho việc thi hành nghĩa vụ pháp lý của tòa án thay đổi theo thời gian, một công 
# ty bảo hiểm muốn so sánh mức trung bình của phần thưởng cho việc thi hành nghĩa vụ pháp lý cá nhân hiện hành với 
# mức của một năm trước đó. Một mẫu ngẫu nhiên gồm n = 30 vụ kiện được chọn lựa trong các vụ kiện được phân xử trong 
# từng năm trong số hai thời kỳ hàng năm này. Các số trung bình và phương sai mẫu của các phần thưởng cho việc thi 
# hành nghĩa vụ pháp lý ( tính bằng triệu đô la Mỹ) cho mỗi trong số hai năm này được cho trong bảng sau :          
#                               Các số trung bình và phương sai mẫu
# Năm	          Cỡ mẫu        |   Trung bình mẫu ( triệu $)                      | Phương sai mẫu ( triệu $)^2
# Hiện tại       n1=30                  X1 = 1.32                                         s1^2=0.9734
# Trước đó       n2 =30                  X2 =1.04                                         s2^2 = 0.7291                                                                
# a) Hãy tìm một ước lượng điểm cho chênh lệch trong mức trung bình về phần thưởng cho việc thi hành nghĩa vụ pháp lý 
# giữa năm hiện tại và năm trước đó. Cho biết biên sai số
# ƒ(X1 - X2) = X1 - X2

# Trong đó:

# X1 là trung bình mẫu cho năm hiện tại (1.32 triệu đô la Mỹ).
# X2 là trung bình mẫu cho năm trước đó (1.04 triệu đô la Mỹ).
# Thay các giá trị vào công thức:

# ƒ(X1 - X2) = 1.32 - 1.04 = 0.28 triệu đô la Mỹ.

# Vậy, ước lượng điểm cho chênh lệch trong mức trung bình về phần thưởng cho việc thi hành nghĩa vụ pháp lý giữa 
# năm hiện tại và năm trước đó là 0.28 triệu đô la Mỹ.

# Để tính toán biên sai số cho ước lượng này, bạn cần sử dụng phương sai mẫu và kích thước mẫu của cả hai 
# năm hiện tại và năm trước đó. Biên sai số (SE) được tính bằng công thức sau:

# SE = sqrt((s1^2/n1) + (s2^2/n2))

# Trong đó:

# s1^2 là phương sai mẫu cho năm hiện tại (0.9734 triệu đô la Mỹ)^2.
# n1 là kích thước mẫu cho năm hiện tại (30).
# s2^2 là phương sai mẫu cho năm trước đó (0.7291 triệu đô la Mỹ)^2.
# n2 là kích thước mẫu cho năm trước đó (30).
# Thay các giá trị vào công thức:

# SE = sqrt((0.9734/30) + (0.7291/30)) = sqrt(0.03245 + 0.02430) = sqrt(0.05675) ≈ 0.2383 triệu đô la Mỹ.

# Vậy, biên sai số cho ước lượng điểm là khoảng ±0.2383 triệu đô la Mỹ.

# b)Tìm một khoảng tin cậy 90% cho chênh lệch trong mức trung bình về phần thưởng cho việc thi hành nghĩa 
# vụ pháp lý giữa năm hiện tại và năm trước đó.

# Khoảng tin cậy = ƒ(Ước lượng điểm) ± (z * SE)

# Trong đó:

# Ước lượng điểm (point estimate) đã tính ở phần trước là 0.28 triệu đô la Mỹ.
# SE (biên sai số) đã tính ở phần trước là khoảng ±0.2307 triệu đô la Mỹ.
# z là z-score liên quan đến mức tin cậy 90%. Đối với mức tin cậy 90%, z-score là khoảng -1.645 đến 1.645.
# Vì chúng ta muốn tính khoảng tin cậy 90%, nên z = 1.645.

# Bây giờ, chúng ta có thể tính khoảng tin cậy:

# Khoảng tin cậy = 0.28 ± (1.645 * 0.2307)

# Khoảng tin cậy = 0.28 ± 0.3796715

# Khoảng tin cậy dự kiến là từ 0.28 - 0.3796715 đến 0.28 + 0.3796715.

# Khoảng tin cậy 90% cho chênh lệch trong mức trung bình về phần thưởng cho việc thi hành nghĩa vụ pháp lý 
# giữa năm hiện tại và năm trước đó là từ khoảng -0.0996715 đến 0.6596715 triệu đô la Mỹ.
import math

# Trung bình mẫu và phương sai mẫu cho năm hiện tại
trungBinhNamHienTai = 1.32  # Trung bình mẫu năm hiện tại (triệu đô la Mỹ)
phuongSaiNamHienTai = 0.9734  # Phương sai mẫu năm hiện tại (triệu đô la Mỹ)^2
kichThuocMauNamHienTai = 30  # Kích thước mẫu năm hiện tại

# Trung bình mẫu và phương sai mẫu cho năm trước đó
trungBinhNamTruocDo = 1.04  # Trung bình mẫu năm trước đó (triệu đô la Mỹ)
phuongSaiNamTruocDo = 0.7291  # Phương sai mẫu năm trước đó (triệu đô la Mỹ)^2
kichThuocMauNamTruocDo = 30  # Kích thước mẫu năm trước đó

# Tính ước lượng điểm cho chênh lệch trong mức trung bình
uocLuongDiem = trungBinhNamHienTai - trungBinhNamTruocDo

# Tính biên sai số (SE)
SE = math.sqrt((phuongSaiNamHienTai / kichThuocMauNamHienTai) + (phuongSaiNamTruocDo / kichThuocMauNamTruocDo))

# In kết quả
print(f"a)")
print(f"Ước lượng điểm cho chênh lệch trong mức trung bình: {uocLuongDiem} triệu đô la Mỹ")
print(f"Biên sai số: ±{SE} triệu đô la Mỹ")

# Z-score cho mức tin cậy 90%
z = 1.645

# Tính khoảng tin cậy 90%
khoangTinCay = (uocLuongDiem - z * SE, uocLuongDiem + z * SE)

# In kết quả
print(f"b)")
print(f"Khoảng tin cậy 90% cho chênh lệch trong mức trung bình: ({khoangTinCay[0]} ; {khoangTinCay[1]})")
```

```python
# Bài 2:
# Một cửa hàng thực phẩm nhận thấy thời gian vừa qua trung bình một khách
# hàng mua 25 nghìn đồng thực phẩm trong ngày. Nay cửa hàng chọn ngẫu nhiên 15
# khách hàng thấy trung bình một khách hàng mua 24 nghìn đồng trong ngày và
# phương sai mẫu điều chỉnh là s2 = (2 nghìn đồng)2. Với mức ý nghĩa là 5% , thử
# xem có phải sức mua của khách hàng hiện nay thực sự giảm sút.

import numpy as np
from scipy import stats

# Dữ liệu
n1 = 15  # Kích thước mẫu hiện tại
mean1 = 24  # Trung bình mẫu hiện tại (triệu đồng)
std1 = 2  # Độ lệch chuẩn mẫu hiện tại (triệu đồng)

# Giả thuyết null: Mức tiêu dùng trung bình hiện tại không thấp hơn so với trước đây
# Giả thuyết thay thế: Mức tiêu dùng trung bình hiện tại thấp hơn so với trước đây
alpha = 0.05  # Mức ý nghĩa

# Tính giá trị t-score
t_score = (mean1 - 25) / (std1 / np.sqrt(n1))

# Tính độ tự do
df = n1 - 1

# Tính giá trị t critical (giá trị t cho alpha/2 và df)
t_critical = stats.t.ppf(1 - alpha, df)

# Tính p-value
p_value = 1 - stats.t.cdf(t_score, df)

# In kết quả
print("Giá trị t-score:", t_score)
print("Giá trị t critical:", t_critical)
print("p-value:", p_value)

# So sánh p-value với mức ý nghĩa alpha
if p_value < alpha:
    print("Có đủ bằng chứng để bác bỏ giả thuyết null.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thuyết null.")
```

```python
# Bài 3:
# Giám đốc một xí nghiệp cho biết lương trung bình của một công nhân thuộc
# xí nghiệp là 380 nghìn đồng/ tháng. Chọn ngẫu nhiên 36 công nhân thấy lương trung
# bình là 350 nghìn đồng/ tháng, với độ lệch chuẩn 𝜎 = 40 nghìn. Lời báo cáo của
# giám đốc có tin cậy được không, với mức ý nghĩa là 5%.

#Giải tay: 
 
import scipy.stats as stats
import math

# Các giá trị đã biết
luongtrungbinhdanso = 380  # Lương trung bình của dân số
luongtrungbinhmau = 350  # Lương trung bình của mẫu
dolechchuanmau = 40  # Độ lệch chuẩn của mẫu
kichthuocmau = 36  # Kích thước mẫu

# Bước 1: Xây dựng giả thuyết
# H0: Lương trung bình của công nhân là 380 nghìn đồng/tháng
# H1: Lương trung bình của công nhân không bằng 380 nghìn đồng/tháng

# Bước 2: Chọn mức ý nghĩa (alpha level)
alpha = 0.05

# Bước 3: Tính giá trị t-score
tscore = (luongtrungbinhmau - luongtrungbinhdanso) / (dolechchuanmau / math.sqrt(kichthuocmau))

# Bước 4: Tính độ tự do và tìm giá trị t critical
sodotudo = kichthuocmau - 1
giatritcritical = stats.t.ppf(1 - alpha/2, sodotudo)  # Với mức ý nghĩa 0.05/2 (kiểm định hai đuôi)

# Bước 5: Tính p-value
pvalue = 2 * (1 - stats.t.cdf(abs(tscore), sodotudo))

# Bước 6: Đưa ra kết luận
if pvalue < alpha:
    print("Bác bỏ giả thuyết null.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thuyết null.")

# In giá trị t-score, t critical và p-value
print("Giá trị t-score:", tscore)
print("Giá trị t critical:", giatritcritical)
print("p-value:", pvalue)
```

```python
# Bài 4:
# Một công ty muốn dự đoán số phút gọi vào tổng đài dịch vụ của khách dựa vào số linh kiện cần sửa chữa, hãy xây dựng mô hình hồi quy theo mô hình sau:
# Minutes = Bo + ß,Units † E,
# Length of Service Calls (in Minutes) and Number of Units Repaired
# Row           |Minutes	|          Units	|    Row | 	      Minutes |      Units
# 1	            23	                1	         8	            97             6
# 2	            29	                2	         9	            109            7
# 3	            49	                3	         10	            119            8
# 4	            64	                4	         11	            149            9
# 5	            74	                4	         12	            145            9
# 6	           87	                5	         13	            154            10
# 7	           96	                6 	         14	            166            10
import numpy as np

# Dữ liệu
units = np.array([1, 2, 3, 4, 4, 5, 6, 6, 7, 8, 9, 9, 10, 10])
minutes = np.array([23, 29, 49, 64, 74, 87, 96, 97, 109, 119, 149, 145, 154, 166])

from sklearn.linear_model import LinearRegression

# Khởi tạo mô hình hồi quy tuyến tính
model = LinearRegression()

# Phù hợp dữ liệu vào mô hình
X = units.reshape(-1, 1)  # Biến độc lập (số lượng linh kiện)
y = minutes  # Biến phụ thuộc (số phút)

model.fit(X, y)  # Huấn luyện mô hình

# Trích xuất hệ số hồi quy (Bo) và hệ số hồi quy (ß)
Bo = model.intercept_
ß = model.coef_[0]

print(f"Bo (Intercept): {Bo}")
print(f"ß (Slope): {ß}")
# Dự đoán số phút cho một số linh kiện cần sửa chữa cụ thể
so_linh_kien = 5
dudoan_phut = model.predict(np.array([[so_linh_kien]]))
print(f"Dự đoán số phút cho {so_linh_kien} linh kiện là: {dudoan_phut[0]} phút")
```
