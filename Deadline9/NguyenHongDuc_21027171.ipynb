```python
# Cau 1: Nháº­n thá»©c Ä‘Æ°á»£c ráº±ng pháº§n thÆ°á»Ÿng cho viá»‡c thi hÃ nh nghÄ©a vá»¥ phÃ¡p lÃ½ cá»§a tÃ²a Ã¡n thay Ä‘á»•i theo thá»i gian, má»™t cÃ´ng 
# ty báº£o hiá»ƒm muá»‘n so sÃ¡nh má»©c trung bÃ¬nh cá»§a pháº§n thÆ°á»Ÿng cho viá»‡c thi hÃ nh nghÄ©a vá»¥ phÃ¡p lÃ½ cÃ¡ nhÃ¢n hiá»‡n hÃ nh vá»›i 
# má»©c cá»§a má»™t nÄƒm trÆ°á»›c Ä‘Ã³. Má»™t máº«u ngáº«u nhiÃªn gá»“m n = 30 vá»¥ kiá»‡n Ä‘Æ°á»£c chá»n lá»±a trong cÃ¡c vá»¥ kiá»‡n Ä‘Æ°á»£c phÃ¢n xá»­ trong 
# tá»«ng nÄƒm trong sá»‘ hai thá»i ká»³ hÃ ng nÄƒm nÃ y. CÃ¡c sá»‘ trung bÃ¬nh vÃ  phÆ°Æ¡ng sai máº«u cá»§a cÃ¡c pháº§n thÆ°á»Ÿng cho viá»‡c thi 
# hÃ nh nghÄ©a vá»¥ phÃ¡p lÃ½ ( tÃ­nh báº±ng triá»‡u Ä‘Ã´ la Má»¹) cho má»—i trong sá»‘ hai nÄƒm nÃ y Ä‘Æ°á»£c cho trong báº£ng sau :          
#                               CÃ¡c sá»‘ trung bÃ¬nh vÃ  phÆ°Æ¡ng sai máº«u
# NÄƒm	          Cá»¡ máº«u        |   Trung bÃ¬nh máº«u ( triá»‡u $)                      | PhÆ°Æ¡ng sai máº«u ( triá»‡u $)^2
# Hiá»‡n táº¡i       n1=30                  X1 = 1.32                                         s1^2=0.9734
# TrÆ°á»›c Ä‘Ã³       n2 =30                  X2 =1.04                                         s2^2 = 0.7291                                                                
# a) HÃ£y tÃ¬m má»™t Æ°á»›c lÆ°á»£ng Ä‘iá»ƒm cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh vá» pháº§n thÆ°á»Ÿng cho viá»‡c thi hÃ nh nghÄ©a vá»¥ phÃ¡p lÃ½ 
# giá»¯a nÄƒm hiá»‡n táº¡i vÃ  nÄƒm trÆ°á»›c Ä‘Ã³. Cho biáº¿t biÃªn sai sá»‘
# Æ’(X1 - X2) = X1 - X2

# Trong Ä‘Ã³:

# X1 lÃ  trung bÃ¬nh máº«u cho nÄƒm hiá»‡n táº¡i (1.32 triá»‡u Ä‘Ã´ la Má»¹).
# X2 lÃ  trung bÃ¬nh máº«u cho nÄƒm trÆ°á»›c Ä‘Ã³ (1.04 triá»‡u Ä‘Ã´ la Má»¹).
# Thay cÃ¡c giÃ¡ trá»‹ vÃ o cÃ´ng thá»©c:

# Æ’(X1 - X2) = 1.32 - 1.04 = 0.28 triá»‡u Ä‘Ã´ la Má»¹.

# Váº­y, Æ°á»›c lÆ°á»£ng Ä‘iá»ƒm cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh vá» pháº§n thÆ°á»Ÿng cho viá»‡c thi hÃ nh nghÄ©a vá»¥ phÃ¡p lÃ½ giá»¯a 
# nÄƒm hiá»‡n táº¡i vÃ  nÄƒm trÆ°á»›c Ä‘Ã³ lÃ  0.28 triá»‡u Ä‘Ã´ la Má»¹.

# Äá»ƒ tÃ­nh toÃ¡n biÃªn sai sá»‘ cho Æ°á»›c lÆ°á»£ng nÃ y, báº¡n cáº§n sá»­ dá»¥ng phÆ°Æ¡ng sai máº«u vÃ  kÃ­ch thÆ°á»›c máº«u cá»§a cáº£ hai 
# nÄƒm hiá»‡n táº¡i vÃ  nÄƒm trÆ°á»›c Ä‘Ã³. BiÃªn sai sá»‘ (SE) Ä‘Æ°á»£c tÃ­nh báº±ng cÃ´ng thá»©c sau:

# SE = sqrt((s1^2/n1) + (s2^2/n2))

# Trong Ä‘Ã³:

# s1^2 lÃ  phÆ°Æ¡ng sai máº«u cho nÄƒm hiá»‡n táº¡i (0.9734 triá»‡u Ä‘Ã´ la Má»¹)^2.
# n1 lÃ  kÃ­ch thÆ°á»›c máº«u cho nÄƒm hiá»‡n táº¡i (30).
# s2^2 lÃ  phÆ°Æ¡ng sai máº«u cho nÄƒm trÆ°á»›c Ä‘Ã³ (0.7291 triá»‡u Ä‘Ã´ la Má»¹)^2.
# n2 lÃ  kÃ­ch thÆ°á»›c máº«u cho nÄƒm trÆ°á»›c Ä‘Ã³ (30).
# Thay cÃ¡c giÃ¡ trá»‹ vÃ o cÃ´ng thá»©c:

# SE = sqrt((0.9734/30) + (0.7291/30)) = sqrt(0.03245 + 0.02430) = sqrt(0.05675) â‰ˆ 0.2383 triá»‡u Ä‘Ã´ la Má»¹.

# Váº­y, biÃªn sai sá»‘ cho Æ°á»›c lÆ°á»£ng Ä‘iá»ƒm lÃ  khoáº£ng Â±0.2383 triá»‡u Ä‘Ã´ la Má»¹.

# b)TÃ¬m má»™t khoáº£ng tin cáº­y 90% cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh vá» pháº§n thÆ°á»Ÿng cho viá»‡c thi hÃ nh nghÄ©a 
# vá»¥ phÃ¡p lÃ½ giá»¯a nÄƒm hiá»‡n táº¡i vÃ  nÄƒm trÆ°á»›c Ä‘Ã³.

# Khoáº£ng tin cáº­y = Æ’(Æ¯á»›c lÆ°á»£ng Ä‘iá»ƒm) Â± (z * SE)

# Trong Ä‘Ã³:

# Æ¯á»›c lÆ°á»£ng Ä‘iá»ƒm (point estimate) Ä‘Ã£ tÃ­nh á»Ÿ pháº§n trÆ°á»›c lÃ  0.28 triá»‡u Ä‘Ã´ la Má»¹.
# SE (biÃªn sai sá»‘) Ä‘Ã£ tÃ­nh á»Ÿ pháº§n trÆ°á»›c lÃ  khoáº£ng Â±0.2307 triá»‡u Ä‘Ã´ la Má»¹.
# z lÃ  z-score liÃªn quan Ä‘áº¿n má»©c tin cáº­y 90%. Äá»‘i vá»›i má»©c tin cáº­y 90%, z-score lÃ  khoáº£ng -1.645 Ä‘áº¿n 1.645.
# VÃ¬ chÃºng ta muá»‘n tÃ­nh khoáº£ng tin cáº­y 90%, nÃªn z = 1.645.

# BÃ¢y giá», chÃºng ta cÃ³ thá»ƒ tÃ­nh khoáº£ng tin cáº­y:

# Khoáº£ng tin cáº­y = 0.28 Â± (1.645 * 0.2307)

# Khoáº£ng tin cáº­y = 0.28 Â± 0.3796715

# Khoáº£ng tin cáº­y dá»± kiáº¿n lÃ  tá»« 0.28 - 0.3796715 Ä‘áº¿n 0.28 + 0.3796715.

# Khoáº£ng tin cáº­y 90% cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh vá» pháº§n thÆ°á»Ÿng cho viá»‡c thi hÃ nh nghÄ©a vá»¥ phÃ¡p lÃ½ 
# giá»¯a nÄƒm hiá»‡n táº¡i vÃ  nÄƒm trÆ°á»›c Ä‘Ã³ lÃ  tá»« khoáº£ng -0.0996715 Ä‘áº¿n 0.6596715 triá»‡u Ä‘Ã´ la Má»¹.
import math

# Trung bÃ¬nh máº«u vÃ  phÆ°Æ¡ng sai máº«u cho nÄƒm hiá»‡n táº¡i
trungBinhNamHienTai = 1.32  # Trung bÃ¬nh máº«u nÄƒm hiá»‡n táº¡i (triá»‡u Ä‘Ã´ la Má»¹)
phuongSaiNamHienTai = 0.9734  # PhÆ°Æ¡ng sai máº«u nÄƒm hiá»‡n táº¡i (triá»‡u Ä‘Ã´ la Má»¹)^2
kichThuocMauNamHienTai = 30  # KÃ­ch thÆ°á»›c máº«u nÄƒm hiá»‡n táº¡i

# Trung bÃ¬nh máº«u vÃ  phÆ°Æ¡ng sai máº«u cho nÄƒm trÆ°á»›c Ä‘Ã³
trungBinhNamTruocDo = 1.04  # Trung bÃ¬nh máº«u nÄƒm trÆ°á»›c Ä‘Ã³ (triá»‡u Ä‘Ã´ la Má»¹)
phuongSaiNamTruocDo = 0.7291  # PhÆ°Æ¡ng sai máº«u nÄƒm trÆ°á»›c Ä‘Ã³ (triá»‡u Ä‘Ã´ la Má»¹)^2
kichThuocMauNamTruocDo = 30  # KÃ­ch thÆ°á»›c máº«u nÄƒm trÆ°á»›c Ä‘Ã³

# TÃ­nh Æ°á»›c lÆ°á»£ng Ä‘iá»ƒm cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh
uocLuongDiem = trungBinhNamHienTai - trungBinhNamTruocDo

# TÃ­nh biÃªn sai sá»‘ (SE)
SE = math.sqrt((phuongSaiNamHienTai / kichThuocMauNamHienTai) + (phuongSaiNamTruocDo / kichThuocMauNamTruocDo))

# In káº¿t quáº£
print(f"a)")
print(f"Æ¯á»›c lÆ°á»£ng Ä‘iá»ƒm cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh: {uocLuongDiem} triá»‡u Ä‘Ã´ la Má»¹")
print(f"BiÃªn sai sá»‘: Â±{SE} triá»‡u Ä‘Ã´ la Má»¹")

# Z-score cho má»©c tin cáº­y 90%
z = 1.645

# TÃ­nh khoáº£ng tin cáº­y 90%
khoangTinCay = (uocLuongDiem - z * SE, uocLuongDiem + z * SE)

# In káº¿t quáº£
print(f"b)")
print(f"Khoáº£ng tin cáº­y 90% cho chÃªnh lá»‡ch trong má»©c trung bÃ¬nh: ({khoangTinCay[0]} ; {khoangTinCay[1]})")
```

```python
# BÃ i 2:
# Má»™t cá»­a hÃ ng thá»±c pháº©m nháº­n tháº¥y thá»i gian vá»«a qua trung bÃ¬nh má»™t khÃ¡ch
# hÃ ng mua 25 nghÃ¬n Ä‘á»“ng thá»±c pháº©m trong ngÃ y. Nay cá»­a hÃ ng chá»n ngáº«u nhiÃªn 15
# khÃ¡ch hÃ ng tháº¥y trung bÃ¬nh má»™t khÃ¡ch hÃ ng mua 24 nghÃ¬n Ä‘á»“ng trong ngÃ y vÃ 
# phÆ°Æ¡ng sai máº«u Ä‘iá»u chá»‰nh lÃ  s2 = (2 nghÃ¬n Ä‘á»“ng)2. Vá»›i má»©c Ã½ nghÄ©a lÃ  5% , thá»­
# xem cÃ³ pháº£i sá»©c mua cá»§a khÃ¡ch hÃ ng hiá»‡n nay thá»±c sá»± giáº£m sÃºt.

import numpy as np
from scipy import stats

# Dá»¯ liá»‡u
n1 = 15  # KÃ­ch thÆ°á»›c máº«u hiá»‡n táº¡i
mean1 = 24  # Trung bÃ¬nh máº«u hiá»‡n táº¡i (triá»‡u Ä‘á»“ng)
std1 = 2  # Äá»™ lá»‡ch chuáº©n máº«u hiá»‡n táº¡i (triá»‡u Ä‘á»“ng)

# Giáº£ thuyáº¿t null: Má»©c tiÃªu dÃ¹ng trung bÃ¬nh hiá»‡n táº¡i khÃ´ng tháº¥p hÆ¡n so vá»›i trÆ°á»›c Ä‘Ã¢y
# Giáº£ thuyáº¿t thay tháº¿: Má»©c tiÃªu dÃ¹ng trung bÃ¬nh hiá»‡n táº¡i tháº¥p hÆ¡n so vá»›i trÆ°á»›c Ä‘Ã¢y
alpha = 0.05  # Má»©c Ã½ nghÄ©a

# TÃ­nh giÃ¡ trá»‹ t-score
t_score = (mean1 - 25) / (std1 / np.sqrt(n1))

# TÃ­nh Ä‘á»™ tá»± do
df = n1 - 1

# TÃ­nh giÃ¡ trá»‹ t critical (giÃ¡ trá»‹ t cho alpha/2 vÃ  df)
t_critical = stats.t.ppf(1 - alpha, df)

# TÃ­nh p-value
p_value = 1 - stats.t.cdf(t_score, df)

# In káº¿t quáº£
print("GiÃ¡ trá»‹ t-score:", t_score)
print("GiÃ¡ trá»‹ t critical:", t_critical)
print("p-value:", p_value)

# So sÃ¡nh p-value vá»›i má»©c Ã½ nghÄ©a alpha
if p_value < alpha:
    print("CÃ³ Ä‘á»§ báº±ng chá»©ng Ä‘á»ƒ bÃ¡c bá» giáº£ thuyáº¿t null.")
else:
    print("KhÃ´ng Ä‘á»§ báº±ng chá»©ng Ä‘á»ƒ bÃ¡c bá» giáº£ thuyáº¿t null.")
```

```python
# BÃ i 3:
# GiÃ¡m Ä‘á»‘c má»™t xÃ­ nghiá»‡p cho biáº¿t lÆ°Æ¡ng trung bÃ¬nh cá»§a má»™t cÃ´ng nhÃ¢n thuá»™c
# xÃ­ nghiá»‡p lÃ  380 nghÃ¬n Ä‘á»“ng/ thÃ¡ng. Chá»n ngáº«u nhiÃªn 36 cÃ´ng nhÃ¢n tháº¥y lÆ°Æ¡ng trung
# bÃ¬nh lÃ  350 nghÃ¬n Ä‘á»“ng/ thÃ¡ng, vá»›i Ä‘á»™ lá»‡ch chuáº©n ğœ = 40 nghÃ¬n. Lá»i bÃ¡o cÃ¡o cá»§a
# giÃ¡m Ä‘á»‘c cÃ³ tin cáº­y Ä‘Æ°á»£c khÃ´ng, vá»›i má»©c Ã½ nghÄ©a lÃ  5%.

#Giáº£i tay: 
 
import scipy.stats as stats
import math

# CÃ¡c giÃ¡ trá»‹ Ä‘Ã£ biáº¿t
luongtrungbinhdanso = 380  # LÆ°Æ¡ng trung bÃ¬nh cá»§a dÃ¢n sá»‘
luongtrungbinhmau = 350  # LÆ°Æ¡ng trung bÃ¬nh cá»§a máº«u
dolechchuanmau = 40  # Äá»™ lá»‡ch chuáº©n cá»§a máº«u
kichthuocmau = 36  # KÃ­ch thÆ°á»›c máº«u

# BÆ°á»›c 1: XÃ¢y dá»±ng giáº£ thuyáº¿t
# H0: LÆ°Æ¡ng trung bÃ¬nh cá»§a cÃ´ng nhÃ¢n lÃ  380 nghÃ¬n Ä‘á»“ng/thÃ¡ng
# H1: LÆ°Æ¡ng trung bÃ¬nh cá»§a cÃ´ng nhÃ¢n khÃ´ng báº±ng 380 nghÃ¬n Ä‘á»“ng/thÃ¡ng

# BÆ°á»›c 2: Chá»n má»©c Ã½ nghÄ©a (alpha level)
alpha = 0.05

# BÆ°á»›c 3: TÃ­nh giÃ¡ trá»‹ t-score
tscore = (luongtrungbinhmau - luongtrungbinhdanso) / (dolechchuanmau / math.sqrt(kichthuocmau))

# BÆ°á»›c 4: TÃ­nh Ä‘á»™ tá»± do vÃ  tÃ¬m giÃ¡ trá»‹ t critical
sodotudo = kichthuocmau - 1
giatritcritical = stats.t.ppf(1 - alpha/2, sodotudo)  # Vá»›i má»©c Ã½ nghÄ©a 0.05/2 (kiá»ƒm Ä‘á»‹nh hai Ä‘uÃ´i)

# BÆ°á»›c 5: TÃ­nh p-value
pvalue = 2 * (1 - stats.t.cdf(abs(tscore), sodotudo))

# BÆ°á»›c 6: ÄÆ°a ra káº¿t luáº­n
if pvalue < alpha:
    print("BÃ¡c bá» giáº£ thuyáº¿t null.")
else:
    print("KhÃ´ng Ä‘á»§ báº±ng chá»©ng Ä‘á»ƒ bÃ¡c bá» giáº£ thuyáº¿t null.")

# In giÃ¡ trá»‹ t-score, t critical vÃ  p-value
print("GiÃ¡ trá»‹ t-score:", tscore)
print("GiÃ¡ trá»‹ t critical:", giatritcritical)
print("p-value:", pvalue)
```

```python
# BÃ i 4:
# Má»™t cÃ´ng ty muá»‘n dá»± Ä‘oÃ¡n sá»‘ phÃºt gá»i vÃ o tá»•ng Ä‘Ã i dá»‹ch vá»¥ cá»§a khÃ¡ch dá»±a vÃ o sá»‘ linh kiá»‡n cáº§n sá»­a chá»¯a, hÃ£y xÃ¢y dá»±ng mÃ´ hÃ¬nh há»“i quy theo mÃ´ hÃ¬nh sau:
# Minutes = Bo + ÃŸ,Units â€  E,
# Length of Service Calls (in Minutes) and Number of Units Repaired
# Row           |Minutes	|          Units	|    Row | 	      Minutes |      Units
# 1	            23	                1	         8	            97             6
# 2	            29	                2	         9	            109            7
# 3	            49	                3	         10	            119            8
# 4	            64	                4	         11	            149            9
# 5	            74	                4	         12	            145            9
# 6	           87	                5	         13	            154            10
# 7	           96	                6 	         14	            166            10
import numpy as np

# Dá»¯ liá»‡u
units = np.array([1, 2, 3, 4, 4, 5, 6, 6, 7, 8, 9, 9, 10, 10])
minutes = np.array([23, 29, 49, 64, 74, 87, 96, 97, 109, 119, 149, 145, 154, 166])

from sklearn.linear_model import LinearRegression

# Khá»Ÿi táº¡o mÃ´ hÃ¬nh há»“i quy tuyáº¿n tÃ­nh
model = LinearRegression()

# PhÃ¹ há»£p dá»¯ liá»‡u vÃ o mÃ´ hÃ¬nh
X = units.reshape(-1, 1)  # Biáº¿n Ä‘á»™c láº­p (sá»‘ lÆ°á»£ng linh kiá»‡n)
y = minutes  # Biáº¿n phá»¥ thuá»™c (sá»‘ phÃºt)

model.fit(X, y)  # Huáº¥n luyá»‡n mÃ´ hÃ¬nh

# TrÃ­ch xuáº¥t há»‡ sá»‘ há»“i quy (Bo) vÃ  há»‡ sá»‘ há»“i quy (ÃŸ)
Bo = model.intercept_
ÃŸ = model.coef_[0]

print(f"Bo (Intercept): {Bo}")
print(f"ÃŸ (Slope): {ÃŸ}")
# Dá»± Ä‘oÃ¡n sá»‘ phÃºt cho má»™t sá»‘ linh kiá»‡n cáº§n sá»­a chá»¯a cá»¥ thá»ƒ
so_linh_kien = 5
dudoan_phut = model.predict(np.array([[so_linh_kien]]))
print(f"Dá»± Ä‘oÃ¡n sá»‘ phÃºt cho {so_linh_kien} linh kiá»‡n lÃ : {dudoan_phut[0]} phÃºt")
```
