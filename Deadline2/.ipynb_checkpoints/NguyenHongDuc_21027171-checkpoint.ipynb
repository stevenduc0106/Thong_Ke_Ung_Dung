```python
#Nguyen Hong Duc 21027171

#Bài 1. Viết chương trình nhập vào độ dài 3 cạnh a, b, c của một tam giác. Tính chu vi và
#diện tích của tam giác theo công thức:
#Chu vi CV = a+b+c
#Diện tích S = sqrt(p*(p-a)*(p-b)*(p-c))
#Trong đó: p=CV/2
#In các kết quả lên màn hình 

import math 

while True:
    # Nhập độ dài các cạnh
    a = float(input("Nhap Do Dai Canh A: ")) 
    b = float(input("Nhap Do Dai Canh B: ")) 
    c = float(input("Nhap Do Dai Canh C: ")) 
    
    # Kiểm tra 3 cạnh có thể tạo thành một tam giác không
    if a + b > c and a + c > b and b + c > a:
        break
    else:
        print("Ba canh này khong the tao thanh mot tam giac. Vui long nhap lai.")
        print()

# Tính chu vi và diện tích tam giác
CV = a + b + c 
p = CV / 2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))
print()
# In kết quả
print("Chu vi cua tam giac: {:.2f}".format(CV))
print("Dien tich cua tam giac: {:.2f}".format(S))
```

```python
#Bài 2. Viết chương trình nhập vào tọa độ của hai điểm (x1, y1) và (x2, y2)
#a) Tính hệ số góc của đường thẳng đi qua hai điểm đó theo công thức:
#Hệ số góc = (y2 - y1) /(x2 - x1)
#b) Tính khoảng cách giữa hai điểm theo công thức:
#Khoảng cách = √(y2 − y1)2 + (x2 − x1)2

import math

#Nhap toa do diem thu nhat
x1 = float (input("Nhap toa do x1: "))
y1 = float (input("Nhap toa do y1: "))

#Nhap toa do diem thu hai
x2 = float (input("Nhap toa do x2: "))
y2 = float (input("Nhap toa do y2: "))

#Tinh he so goc
HSG = (y2 - y1) / (x2 - x1)

#Tinh khoang cach
KC = math.sqrt( ( y2 - y1 )**2 + ( x2 - x1 )**2 )

#In ket qua
print("He So Goc = {:.2f}".format(HSG))
print("Khoang Cach = {:.2f}".format(KC))
```

```python
#Bài 3. Trò chơi "Oẳn tù tì": trò chơi có 2 người chơi mỗi người sẽ dùng tay để biểu thị một
#trong 3 công cụ sau: Kéo, Bao và Búa.
#Nguyên tắc: Kéo thắng bao.
#Bao thắng búa.
#Búa thắng kéo.
#Viết chương trình mô phỏng trò chơi này cho hai người chơi và người chơi với máy

import random

# Ham de lay lua chon tu nguoi choi hoac may
def layLuaChon(playerName):
    while True:
        choice = input(f"{playerName}, nhap lua chon cua ban (1-Keo / 2-Bua / 3-Bao): ")
        if choice in ['1', '2', '3']:
            return choice
        else:
            print("Lua chon khong hop le. Vui long chon lai.")

# Ham de xac dinh nguoi thang
def xacDinhNguoiThang(choice1, choice2):
    if choice1 == choice2:
        return "Hoa"
    if (choice1 == '1' and choice2 == '3') or \
       (choice1 == '2' and choice2 == '1') or \
       (choice1 == '3' and choice2 == '2'):
        return "Nguoi choi 1"
    else:
        return "Nguoi choi 2"

# Tao danh sach cac lua chon
choices = ['1', '2', '3']

while True:
    # Hien thi menu chon che do choi
    print("Menu:")
    print("1. Nguoi choi voi nguoi")
    print("2. Nguoi choi voi may")
    print("3. Ket thuc tro choi")
    
    while True:
        mode = input("Chon che do (1, 2 hoac 3): ")
        if mode in ['1', '2', '3']:
            break
        else:
            print("Lua chon khong hop le. Vui long chon lai.")
    
    if mode == '3':
        print("Tro choi da ket thuc.")
        break
    print()
    # Lua chon che do choi
    if mode == '1':
        player1Choice = layLuaChon("Nguoi choi 1")
        player2Choice = layLuaChon("Nguoi choi 2")
    else:
        player1Choice = layLuaChon("Nguoi choi")
        player2Choice = random.choice(choices)

    # In lua chon cua nguoi choi 1 va nguoi choi 2 (hoac may)
    print(f"Nguoi choi 1 chon: {choices[int(player1Choice) - 1]}")
    print(f"Nguoi choi 2 (hoac may) chon: {choices[int(player2Choice) - 1]}")
    print()
    # Xac dinh nguoi thang
    winner = xacDinhNguoiThang(player1Choice, player2Choice)
    if winner == "Hoa":
        print("Ket qua: Hoa\n")
    else:
        print(f"Ket qua: {winner} thang!\n")
```

```python
#Bài 4. Viết chương trình tính tiền điện gồm các khoản sau:
#Tiền thuê bao điện kế : 1000 đồng / tháng.
#Định mức sử dụng điện cho mỗi hộ là 50 Kw
#Phần định mức tính giá 450 đồng /Kwh
#Nếu phần vượt định mức <= 50 Kw tính giá phạt cho phần này là 700 đồng/Kwh .
#Nếu phần vượt định mức lớn 50 Kw và nhỏ hơn 100Kw tính giá phạt cho phần này
#là 910 đồng/Kwh
#Nếu phần vượt định mức lớn hơn hay bằng 100 Kw tính giá phạt cho phần này là
#1200 đồng/Kwh .
#Với : chỉ số điện kế cũ và chỉ số điện kế mới nhập vào từ bàn phím. In ra màn hình
#số tiền trả trong định mức, vượt định mức và tổng của chúng. 

#Nhap chi so đien cu va moi
while True:
    chi_so_cu = float(input("Nhap chi so dien cu: "))
    chi_so_moi = float(input("Nhap chi so dien moi: "))
    print()
    if chi_so_moi > chi_so_cu:
        break
    else:
        print("Chi so dien moi phai lon hon chi so dien cu. Vui long nhap lai.")
        print()

so_luong_kw = chi_so_moi - chi_so_cu

#Tien thue bao dien co đinh
tien_thue_bao = 1000

#Dinh muc su dung dien cho moi ho la 50 Kw
dinh_muc = 50

#Phan vuot dinh muc, neu la so am thi coi la 0
vuot_dinh_muc = max(0, so_luong_kw - dinh_muc)

#Tinh tien trong dinh muc
tien_trong_dinh_muc = dinh_muc * 450

#Tinh tien vuot dinh muc theo tung muc gia
if vuot_dinh_muc <= 50:
    tien_vuot_dinh_muc = vuot_dinh_muc * 700
elif vuot_dinh_muc <= 100:
    tien_vuot_dinh_muc = 50 * 700 + (vuot_dinh_muc - 50) * 910
else:
    tien_vuot_dinh_muc = 50 * 700 + 50 * 910 + (vuot_dinh_muc - 100) * 1200

#Tinh tong tien dien
tong_tien = tien_thue_bao + tien_trong_dinh_muc + tien_vuot_dinh_muc

#In ket qua
print("So tien trong đinh muc:", "{:,.2f} VND".format(tien_trong_dinh_muc))
print("So tien vuot đinh muc:", "{:,.2f} VND".format(tien_vuot_dinh_muc))
print("Tong so tien:", "{:,.2f} VND".format(tong_tien))
```

```python
#Bài 5. Viết chương trình tính các tổng sau:
#a) S=1/2 + 2/3 + ... +n/(n+1)
#Nhap n
n = int(input("a) Nhap n: "))

sum_a = 0

for i in range(1, n + 1):
    sum_a += i / (i + 1)

#In ket qua
print("Tổng S = {:.2f}".format(sum_a))

# ----------------------------------------------

#b) S= - 1 +2 - 3 +4 - ... + (-1)^n *n 
#Nhap n
while True:
    n = int(input("Nhap n: "))
    if n > 0:
        break
    else:
        print("Vui long nhap n la mot so nguyen duong.")

sum_b = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        sum_b += i
    else:
        sum_b -= i

#In ket qua
print("Tong S = {:.2f}".format(sum_b))
```

```python
#Bài 6. Fibonacci là một dãy số được định nghĩa như sau:
#F
#n
#= {
#1, n = 1
#2, n＝２
#Fn−1 + Fn−2
#, n > 2
#Viết chương trình in ra màn hình dãy Fibonacci có n số hạng, n nhập từ bàn phím
#khi cho chạy chương trình. 

#Nhap n
n = int(input("Nhập so luong so hang n: "))

a, b = 1, 2

#In hai so hang dau tien
print(a, b, end=" ")

#Tinh day fibonacci
for _ in range(2, n):
    F = a + b
    print(F, end=" ")
    
    # Cap nhat a và b de tinh so hang ke tiep
    a, b = b,F
```
