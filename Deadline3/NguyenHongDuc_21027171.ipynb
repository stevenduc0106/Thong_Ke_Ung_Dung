```python
#3- Viết hàm tính giá trị lớn nhất, nhỏ nhất và trung bình (tự viết code không sử dụng thư
#viện).

def tinhGiaTriMaxMinTrungBinh(danhSach):
    if len(danhSach) == 0:
        return None, None, None

    max_value = danhSach[0]
    min_value = danhSach[0]
    total = 0

    for value in danhSach:
        if value > max_value:
            max_value = value
        if value < min_value:
            min_value = value
        total += value

    mean_value = total / len(danhSach)

    return max_value, min_value, mean_value

#4- Áp dụng 3 hàm vừa viết ở 3 vào từng cột 1,2,3,4 và hiển thị kết quả ra màn hình

with open("iris.data", "r") as file:
    dulieu = []
    dong = file.readline()
    while dong:
        dulieu.append(dong.strip())
        dong = file.readline()

cot1 = []
cot2 = []
cot3 = []
cot4 = []

for dong in dulieu:
    thuocTinh = dong.strip().split(",")
    if len(thuocTinh) == 5:
        cot1.append(float(thuocTinh[0]))
        cot2.append(float(thuocTinh[1]))
        cot3.append(float(thuocTinh[2]))
        cot4.append(float(thuocTinh[3]))

#Tinh Max, Min, Mean
cot1Max, cot1Min, cot1Avg = tinhGiaTriMaxMinTrungBinh(cot1)
cot2Max, cot2Min, cot2Avg = tinhGiaTriMaxMinTrungBinh(cot2)
cot3Max, cot3Min, cot3Avg = tinhGiaTriMaxMinTrungBinh(cot3)
cot4Max, cot4Min, cot4Avg = tinhGiaTriMaxMinTrungBinh(cot4)

print("{:<20} {:<20} {:<20} {:<20}".format("Thuoc tinh", "Gia tri MAX", "Gia tri MIN", "Gia tri MEAN"))

print("{:<20} {:>11.2f} {:>20.2f} {:>20.2f}".format("sepal length", cot1Max, cot1Min, cot1Avg))
print("{:<20} {:>11.2f} {:>20.2f} {:>20.2f}".format("sepal width", cot2Max, cot2Min, cot2Avg))
print("{:<20} {:>11.2f} {:>20.2f} {:>20.2f}".format("petal length", cot3Max, cot3Min, cot3Avg))
print("{:<20} {:>11.2f} {:>20.2f} {:>20.2f}".format("petal width", cot4Max, cot4Min, cot4Avg))

```

```python
#7- Viết hàm tính mối độ tương quan pearson theo công thức sau:

def tinhMoiDoTuongQuanPearson(x, y):
    trungBinhX = sum(x) / len(x)
    trungBinhY = sum(y) / len(y)
    
    tuSo = sum((x[i] - trungBinhX) * (y[i] - trungBinhY) for i in range(len(x)))
    mauSoX = sum((x[i] - trungBinhX) ** 2 for i in range(len(x)))
    mauSoY = sum((y[i] - trungBinhY) ** 2 for i in range(len(y)))
    
    doLechChuanX = (mauSoX / (len(x) - 1)) ** 0.5
    doLechChuanY = (mauSoY / (len(y) - 1)) ** 0.5
    
    moiDoTuongQuan = tuSo / (doLechChuanX * doLechChuanY)
    
    return moiDoTuongQuan

x = [1, 2, 3, 4, 8]
y = [2, 3, 4, 5, 6]

moiDoTuongQuan = tinhMoiDoTuongQuanPearson(x, y)
print("Mối độ tương quan Pearson: {:.4f}".format(moiDoTuongQuan))
```
