```python
# Bài 1:
# Có 30 đề thi thống kê ứng dụng trong đó có 10 đề khó, 20 đề trung bình. Tìm xác suất để:
# a) Một sinh viên bắt một đề gặp được đề trung bình.
# b) Một sinh viên bắt hai đề, được ít nhất một đề trung bình.

# a)xác suất để một sinh viên chọn ngẫu nhiên một đề gặp được đề trung bình là:
# Xác suất = (Số đề trung bình) / (Tổng số đề) = 20 / 30 = 2/3 ≈ 0.6667

# b)
# Trường hợp 1: Chọn một đề trung bình và một đề khó.

# Xác suất chọn một đề trung bình từ 20 đề trung bình: C(20, 1) = 20
# Xác suất chọn một đề khó từ 10 đề khó: C(10, 1) = 10
# Tổng số cách chọn 2 đề từ 30 đề: C(30, 2) = 435

# Xác suất cho trường hợp này là: (20 * 10) / 435 = 200 / 435.

# Trường hợp 2: Chọn cả hai đề trung bình.

# Xác suất chọn cả hai đề trung bình từ 20 đề trung bình: C(20, 2) = 190
# Tổng số cách chọn 2 đề từ 30 đề: C(30, 2) = 435

# Xác suất cho trường hợp này là: (190) / 435.

# Tổng xác suất của cả hai trường hợp là:

# (200 / 435) + (190 / 435) = 390 / 435 ≈ 0.8966
from math import comb

# Số đề trung bình và đề khó
soDeTrungBinh = 20
soDeKho = 10

# Tổng số đề
tongSoDe = soDeTrungBinh + soDeKho

# a) Xác suất để chọn một đề trung bình
xacSuatA = soDeTrungBinh / tongSoDe

# b) Tính xác suất cho trường hợp 1 (chọn một đề trung bình và một đề khó)
xacSuatTruongHop1 = (comb(soDeTrungBinh, 1) * comb(soDeKho, 1)) / comb(tongSoDe, 2)

# Tính xác suất cho trường hợp 2 (chọn cả hai đề trung bình)
xacSuatTruongHop2 = comb(soDeTrungBinh, 2) / comb(tongSoDe, 2)

# Tổng xác suất của cả hai trường hợp
tongXacSuatB = xacSuatTruongHop1 + xacSuatTruongHop2

print("a) Xác suất để chọn một đề trung bình: {:.4f}".format(xacSuatA))
print("b) Xác suất để chọn ít nhất một đề trung bình: {:.4f}".format(tongXacSuatB))

```

```python
# Bài 2:
# Có hai lớp CNTTA và CNTT B mỗi lớp có 45 học sinh, số sinh viên giỏi Ngoại ngữ và số sinh viên giỏi Giải tích được 
# cho trong bảng sau. Có một đoàn thanh tra. Hiệu trưởng nên mời vào lớp nào để khả năng gặp được một em giỏi ít nhất 
# một môn là cao nhất?

# Lớp CNTTA:

# Số học sinh giỏi ngoại ngữ: 25
# Số học sinh giỏi giải tích: 30
# Số học sinh giỏi cả ngoại ngữ và giải tích: 20
# Lớp CNTTB:

# Số học sinh giỏi ngoại ngữ: 25
# Số học sinh giỏi giải tích: 30
# Số học sinh giỏi cả ngoại ngữ và giải tích: 10
    
# Xác suất một học sinh ở lớp CNTTA giỏi ít nhất một môn là:
# P(CNTTA) = 25/45 + 30/45 - 20/45 = 35/45
# Xác suất một học sinh ở lớp CNTTB giỏi ít nhất một môn là:
# P(CNTTB) =  = 25/45 + 30/45 - 10/45 = 45/45 = 1

# Vậy ta có thể so sánh P(CNTTA) với P(CNTTB) thì P(CNTTB) lớn hơn P(CNTTA) nên hiệu trưởng nên vào lớp CNTT lớp B để
# gặp được một em giỏi ít nhất một môn.

soHocSinhMoiLop = 45

soHocSinhGioiNgoaiNguCNTTA = 25
soHocSinhGioiGiaiTichCNTTA = 30
soHocSinhGioiCaNgoaiNguVaGiaiTichCNTTA = 20

soHocSinhGioiNgoaiNguCNTTB = 25
soHocSinhGioiGiaiTichCNTTB = 30
soHocSinhGioiCaNgoaiNguVaGiaiTichCNTTB = 10

xacSuatCNTTA = (soHocSinhGioiNgoaiNguCNTTA + soHocSinhGioiGiaiTichCNTTA - soHocSinhGioiCaNgoaiNguVaGiaiTichCNTTA) / soHocSinhMoiLop

xacSuatCNTTB = (soHocSinhGioiNgoaiNguCNTTB + soHocSinhGioiGiaiTichCNTTB - soHocSinhGioiCaNgoaiNguVaGiaiTichCNTTB) / soHocSinhMoiLop

print("Xác suất một học sinh ở lớp CNTTA giỏi ít nhất một môn:", xacSuatCNTTA)
print("Xác suất một học sinh ở lớp CNTTB giỏi ít nhất một môn:", xacSuatCNTTB)

if xacSuatCNTTA > xacSuatCNTTB:
    print("Hiệu trưởng nên mời vào lớp CNTTA để khả năng gặp được ít nhất một em giỏi ít nhất một môn cao nhất.")
else:
    print("Hiệu trưởng nên mời vào lớp CNTTB để khả năng gặp được ít nhất một em giỏi ít nhất một môn cao nhất.")
```

```python
# Bài 3:
# Lớp có 100 Sinh viên, trong đó có 50 SV giỏi Anh Văn, 45 SV giỏi Pháp Văn, 10 SV giỏi cả hai ngoại ngữ. Chọn ngẫu nhiên một sinh viên trong lớp. Tính xác suất:
# a) Sinh viên này giỏi ít nhất một ngoại ngữ.
# b) Sinh viên này không giỏi ngoại ngữ nào hết. 
# c) Sinh viên này chỉ giỏi đúng một ngoại ngữ.
# d) Sinh viên này chỉ giỏi duy nhất môn Anh Văn.

# a) Xác suất sinh viên này giỏi ít nhất một ngoại ngữ:
# Xác suất giỏi ít nhất một ngoại ngữ = Xác suất giỏi Anh Văn + Xác suất giỏi Pháp Văn - Xác suất giỏi cả hai ngoại ngữ
# = (50/100) + (45/100) - (10/100) = 0.85

# b) Xác suất sinh viên này không giỏi ngoại ngữ nào hết (không giỏi Anh Văn và cũng không giỏi Pháp Văn):
# Xác suất không giỏi ngoại ngữ nào hết = 1 - Xác suất giỏi ít nhất một ngoại ngữ
# = 1 - 0.85 = 0.15

# c) Xác suất sinh viên này chỉ giỏi đúng một ngoại ngữ (chỉ giỏi Anh Văn hoặc chỉ giỏi Pháp Văn):
# Xác suất chỉ giỏi Anh Văn = (50/100) - (10/100) = 0.40
# Xác suất chỉ giỏi Pháp Văn = (45/100) - (10/100) = 0.35
# Xác suất chỉ giỏi đúng một ngoại ngữ = Xác suất chỉ giỏi Anh Văn + Xác suất chỉ giỏi Pháp Văn
# = 0.40 + 0.35 = 0.75

# d) Xác suất sinh viên này chỉ giỏi duy nhất môn Anh Văn:
# Xác suất chỉ giỏi duy nhất môn Anh Văn = Xác suất giỏi Anh Văn - Xác suất giỏi cả hai ngoại ngữ
# = (50/100) - (10/100) = 0.4
def tinhXacSuat(totalSinhVien, sinhVienGioiAnhVan, sinhVienGioiPhapVan, sinhVienGioiCaHai):
    xacSuatGioiAnhVan = sinhVienGioiAnhVan / totalSinhVien
    xacSuatGioiPhapVan = sinhVienGioiPhapVan / totalSinhVien
    xacSuatGioiCaHai = sinhVienGioiCaHai / totalSinhVien

    xacSuatItNhatMotNgoaiNguGioi = xacSuatGioiAnhVan + xacSuatGioiPhapVan - xacSuatGioiCaHai
    xacSuatKhongGioiNgoaiNgu = 1 - xacSuatItNhatMotNgoaiNguGioi

    xacSuatChiGioiMotNgoaiNgu = (sinhVienGioiAnhVan - sinhVienGioiCaHai) / totalSinhVien + (sinhVienGioiPhapVan - sinhVienGioiCaHai) / totalSinhVien

    xacSuatChiGioiAnhVan = (sinhVienGioiAnhVan - sinhVienGioiCaHai) / totalSinhVien

    return xacSuatItNhatMotNgoaiNguGioi, xacSuatKhongGioiNgoaiNgu, xacSuatChiGioiMotNgoaiNgu, xacSuatChiGioiAnhVan

# Dữ liệu đã cho
totalSinhVien = 100
sinhVienGioiAnhVan = 50
sinhVienGioiPhapVan = 45
sinhVienGioiCaHai = 10

# Tính xác suất
xacSuatItNhatMotNgoaiNguGioi, xacSuatKhongGioiNgoaiNgu, xacSuatChiGioiMotNgoaiNgu, xacSuatChiGioiAnhVan = tinhXacSuat(totalSinhVien, sinhVienGioiAnhVan, sinhVienGioiPhapVan, sinhVienGioiCaHai)

print("a) Xác suất sinh viên này giỏi ít nhất một ngoại ngữ:", xacSuatItNhatMotNgoaiNguGioi)
print("b) Xác suất sinh viên này không giỏi ngoại ngữ nào hết: {:.2f}".format(xacSuatKhongGioiNgoaiNgu))
print("c) Xác suất sinh viên này chỉ giỏi đúng một ngoại ngữ:", xacSuatChiGioiMotNgoaiNgu)
print("d) Xác suất sinh viên này chỉ giỏi duy nhất môn Anh Văn:", xacSuatChiGioiAnhVan)
```

```python
# Bài 4:
# Trong một hộp có 12 bóng đèn, trong đó có 3 bóng hỏng. Lấy ngẫu nhiên không hoàn lại ba bóng để dùng. Tính xác suất để:
# a) Cả ba bóng đều hỏng.
# b) Cả ba bóng đều không hỏng?
# c) Có ít nhất một bóng không hỏng?
# d) Chỉ có bóng thứ hai hỏng?

# a) xác suất cả ba bóng đều hỏng: C(3,3) / C(12,3) = 1/220 ~ 0.004545
# b) xác suất cả ba bóng đều không hỏng: C(9,3) / C(12,3) = 84/220 ~ 0.381818
# c) xác suất có ít nhất một bóng không hỏng: 1 - kq câu a ~ 0.995455
# d) xác suất chỉ có bóng thứ hai hỏng : C(3,1) * C(9,2) / C(12,3) = 0.490909
from math import comb

def tinhXacSuat(tongSoBongDen, bongHong, soBongChon):
    xacSuatTatCaHong = comb(bongHong, soBongChon) / comb(tongSoBongDen, soBongChon)
    xacSuatTatCaHoatDong = comb(tongSoBongDen - bongHong, soBongChon) / comb(tongSoBongDen, soBongChon)
    xacSuatItNhatMotHoatDong = 1 - xacSuatTatCaHong
    xacSuatChiCoBongThuHaiHong = (comb(bongHong, 1) * comb(tongSoBongDen - bongHong, soBongChon - 1)) / comb(tongSoBongDen, soBongChon)

    return xacSuatTatCaHong, xacSuatTatCaHoatDong, xacSuatItNhatMotHoatDong, xacSuatChiCoBongThuHaiHong

# Dữ liệu cho trường hợp
tongSoBongDen = 12
bongHong = 3
soBongChon = 3

# Tính toán xác suất
xacSuatTatCaHong, xacSuatTatCaHoatDong, xacSuatItNhatMotHoatDong, xacSuatChiCoBongThuHaiHong = tinhXacSuat(tongSoBongDen, bongHong, soBongChon)

print("a) Xác suất cả ba bóng đều hỏng: {:.6f}".format(xacSuatTatCaHong))
print("b) Xác suất cả ba bóng đều không hỏng: {:.6f}".format(xacSuatTatCaHoatDong))
print("c) Xác suất có ít nhất một bóng không hỏng: {:.6f}".format(xacSuatItNhatMotHoatDong))
print("d) Xác suất chỉ có bóng thứ hai hỏng: {:.6f}".format(xacSuatChiCoBongThuHaiHong))
```

```python
# Bài 5:
# Một sọt Cam có 10 trái trong đó có 4 trái hư. Lấy ngẫu nhiên ra ba trái.
# a) Tính xác suất lấy được 3 trái hư.
# b) Tính xác suất lấy được 1 trái hư.
# c) Tính xác suất lấy được ít nhất một trái hư.
# d) Tính xác suất lấy được nhiều nhất 2 trái hư.

# a) Xác suất lấy được 3 trái hư:
# chọn 3 trái hư từ 4 trái hư là C(4,3) và tổng số cách chọn 3 trái cam bất kỳ từ tổng số 10 trái cam là C(10,3).
# Xác suất = C(4,3) / C(10,3) = 4 / 120 = 1 / 30 ≈ 0.0333
# b) Tính xác suất lấy được 1 trái hư.
# Số cách chọn 1 trái hư từ 4 trái hư là C(4,1), và số cách chọn 2 trái không hư từ 6 trái không hư còn lại là C(6,2).
# Xác suất = (C(4,1) * C(6,2)) / C(10,3) = 1 / 2 = 0.5
# c) Tính xác suất lấy được ít nhất một trái hư.
# Xác suất không có trái hư nào = C(6,3) / C(10,3) = 20 / 120 = 1 / 6
# Xác suất có ít nhất 1 trái hư = 1 - Xác suất không có trái hư nào = 1 - 1 / 6 = 5 / 6 ≈ 0.8333
# d) Xác suất lấy được nhiều nhất 2 trái hư:
# Xác suất có nhiều nhất 2 trái hư = 1 - Kết quả của câu a = 1-0.033 = 0,967

from math import comb

def tinhXacSuat(tongSoTraiCam, traiCamHu, soTraiChon):
    xacSuat3TraiHu = (comb(traiCamHu, 3) * comb(tongSoTraiCam - traiCamHu, soTraiChon - 3)) / comb(tongSoTraiCam, soTraiChon)
    
    xacSuat1TraiHu = (comb(traiCamHu, 1) * comb(tongSoTraiCam - traiCamHu, soTraiChon - 1)) / comb(tongSoTraiCam, soTraiChon)
    
    xacSuatItNhatMotHu = 1 - (comb(tongSoTraiCam - traiCamHu, soTraiChon) / comb(tongSoTraiCam, soTraiChon))
    
    xacSuatNhieuNhat2Hu = 1 -  xacSuat3TraiHu
    
    return xacSuat3TraiHu, xacSuat1TraiHu, xacSuatItNhatMotHu, xacSuatNhieuNhat2Hu

# Dữ liệu cho bài toán
tongSoTraiCam = 10
traiCamHu = 4
soTraiChon = 3

# Tính toán xác suất
xacSuat3TraiHu, xacSuat1TraiHu, xacSuatItNhatMotHu, xacSuatNhieuNhat2Hu = tinhXacSuat(tongSoTraiCam, traiCamHu, soTraiChon)

# In kết quả
print("a) Xác suất lấy được 3 trái hư: {:.6f}".format(xacSuat3TraiHu))
print("b) Xác suất lấy được 1 trái hư: {:.6f}".format(xacSuat1TraiHu))
print("c) Xác suất lấy được ít nhất một trái hư: {:.6f}".format(xacSuatItNhatMotHu))
print("d) Xác suất lấy được nhiều nhất 2 trái hư: {:.6f}".format(xacSuatNhieuNhat2Hu))
```

```python
# Bài 6:
# Một gia đình có 10 người con. Giả sử xác suất sinh con trai, con gái như nhau. Tính xác suất:
# a) Không có con trai.
# b) Có 5 con trai và 5 con gái.
# c) Số trai từ 5 đến 7.

# a) Không có con trai
# Xác suất sinh con trai là 1/2 và xác suất sinh con gái cũng là 1/2. 
# Do đó, xác suất để có một gia đình với 10 người con đều là con gái là:
# (1/2) * (1/2) * (1/2) * ... * (1/2) = (1/2)^10
# Xác suất = (1/2)^10 = 1/1024 ≈ 0.000977
# b) Có 5 con trai và 5 con gái.
# Số cách có thể chọn 5 đứa con trai và con gái là C(10,5)
# Số cách có thể chọn đứa con là 2^10 
# Xác suất = C(10,5) / 2^10 ~ 0.246094
# c) Số trai từ 5 đến 7.
# Số cách có thể chọn 5, 6 hoặc 7 đứa con là con trai:
# C(10,5) + C(10,6) + C(10,7) = 582
# Số cách tổng cộng có thể chọn 10 đứa con là: 2^10
# Xác suất = 582 / 2^10 ~ 0.568359

from math import comb

def tinhXacSuat(tongSoCon, xacSuatConTrai, soConTraiChon):
    def tinhToHop(a, b):
        return factorial(a) // (factorial(b) * factorial(a - b))
    
    def tinhXacSuat1LanThu(soConTraiChon, tongSoCon, xacSuatConTrai):
        return tinhToHop(tongSoCon, soConTraiChon) * (xacSuatConTrai ** soConTraiChon) * ((1 - xacSuatConTrai) ** (tongSoCon - soConTraiChon))

    def factorial(x):
        if x == 0:
            return 1
        return x * factorial(x - 1)
    
    xacSuat = tinhXacSuat1LanThu(soConTraiChon, tongSoCon, xacSuatConTrai)
    return xacSuat

tongSoCon = 10
xacSuatConTrai = 0.5

xacSuatA = tinhXacSuat(tongSoCon, xacSuatConTrai, 0)
print("a) Xác suất không có con trai: {:.6f}".format(xacSuatA))

xacSuatB = tinhXacSuat(tongSoCon, xacSuatConTrai, 5)
print("b) Xác suất có 5 con trai và 5 con gái: {:.6f}".format(xacSuatB))

xacSuatC = sum([tinhXacSuat(tongSoCon, xacSuatConTrai, k) for k in range(5, 8)])
print("c) Xác suất có từ 5 đến 7 con trai: {:.6f}".format(xacSuatC))
```
