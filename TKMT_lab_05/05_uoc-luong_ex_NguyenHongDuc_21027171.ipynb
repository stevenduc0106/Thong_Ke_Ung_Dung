### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import random
import numpy as np
import statistics
import matplotlib.pyplot as plt
import scipy.stats as stats

# tạo dữ liệu mẫu với 10000 phần tử
heights = [random.randint(120, 200) for _ in range(10000)]

# lưu dữ liệu vào biến POP
POP = heights

# tính chiều cao trung bình của quần thể
mean_height = statistics.mean(POP)
print("Chiều cao trung bình của quần thể là:", mean_height)

# tính độ lệch chuẩn về chiều cao của quần thể
stdev_height = statistics.stdev(POP)
print("Độ lệch chuẩn về chiều cao của quần thể là:", stdev_height)

# số lượng người cao (chiều cao >= 180)
num_tall = len([h for h in POP if h >= 180])

# tỷ lệ người cao trong quần thể (phần trăm)
percent_tall = (num_tall / len(POP)) * 100

print("Tỷ lệ người cao trong quần thể là: {:.2f}%".format(percent_tall))

# Vẽ histogram
plt.hist(POP, bins=50)
plt.xlabel('Chiều cao')
plt.ylabel('Số lượng người')
plt.title('Phân phối chiều cao trong quần thể')
plt.show()

# tính kỳ vọng và độ lệch chuẩn của quần thể
mu = statistics.mean(POP)
sigma = statistics.stdev(POP)

# tạo dữ liệu mẫu cho phân phối chuẩn
norm_data = stats.norm.rvs(loc=mu, scale=sigma, size=10000)

# vẽ histogram của chiều cao
plt.hist(POP, bins=50, density=True, label='Quần thể')

# vẽ đồ thị hàm mật độ phân phối chuẩn
x_range = np.linspace(min(POP), max(POP), 100)
density = stats.norm.pdf(x_range, mu, sigma)
plt.plot(x_range, density, label='Phân phối chuẩn')

plt.xlabel('Chiều cao')
plt.ylabel('Tần suất')
plt.title('Phân phối chiều cao trong quần thể và phân phối chuẩn')
plt.legend()
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
#a)
import random
import matplotlib.pyplot as plt
import statistics

# Khởi tạo mảng để lưu trữ 100 trung bình mẫu
sample_means_a = []

# a) Thực hiện Bài 2 100 lần
for _ in range(100):
    # Lấy mẫu kích thước 20
    sample = random.sample(POP, 20)
    
    # Tính trung bình mẫu
    sample_mean = statistics.mean(sample)
    
    # Thêm trung bình mẫu vào mảng
    sample_means_a.append(sample_mean)

# Vẽ histogram cho 100 trung bình mẫu
plt.hist(sample_means_a, bins=20)

plt.xlabel('Trung bình mẫu')
plt.ylabel('Số lần xuất hiện')
plt.title('Phân phối của 100 trung bình mẫu (Kích thước mẫu = 20)')
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
import random
import matplotlib.pyplot as plt
import statistics

# Khởi tạo danh sách để lưu trữ phương sai của các mẫu
variance_list = []

# Thực hiện lấy mẫu 1000 lần
for _ in range(1000):
    # Lấy mẫu kích thước 100
    sample = random.sample(POP, 100)
    
    # Tính độ lệch chuẩn của mẫu
    sample_std = statistics.stdev(sample)
    
    # Tính phương sai của mẫu (bình phương của độ lệch chuẩn)
    sample_variance = sample_std ** 2
    
    # Thêm phương sai vào danh sách
    variance_list.append(sample_variance)

# Vẽ histogram của phương sai của các mẫu
plt.hist(variance_list, bins=20)

plt.xlabel('Phương sai mẫu')
plt.ylabel('Số lần xuất hiện')
plt.title('Phân phối phương sai của các mẫu (Kích thước mẫu = 100)')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import math
import scipy.stats as stats

#Mẫu quan sát
x = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]

#Kích thước của mẫu
n = len(x)

#Độ tin cậy
alpha = 0.05

#Giá trị T-score cho độ tin cậy 95% với (n-1) = 8 độ tự do
t = stats.t.ppf(1 - alpha / 2, n - 1)

#Độ lệch chuẩn của quần thể
sigma = 0.3

#Giá trị trung bình mẫu
avg = sum(x)/n

#Khoảng tin cậy cho trung bình mẫu khi sigma đã biết
LB = avg - t * (sigma/math.sqrt(n))
UB = avg + t * (sigma/math.sqrt(n))

print("Khoảng tin cậy cho cân nặng trung bình của gà khi sigma = 0.3:")
print("LB =", round(LB,3))
print("UB =", round(UB,3))
```

```python
import math
import scipy.stats as stats

# Mẫu quan sát
x = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]

# Kích thước của mẫu
n = len(x)

# Độ tin cậy
alpha = 0.05

# Giá trị T-score cho độ tin cậy 95% với (n-1) = 8 độ tự do
t = stats.t.ppf(1 - alpha / 2, n - 1)

# Độ lệch chuẩn của mẫu
s = statistics.stdev(x)

# Giá trị trung bình mẫu
avg = sum(x)/n

# Khoảng tin cậy cho trung bình mẫu
LB = avg - t * (s/math.sqrt(n))
UB = avg + t * (s/math.sqrt(n))

print("Khoảng tin cậy cho cân nặng trung bình của gà khi không biết sigma:")
print("LB =", round(LB,3))
print("UB =", round(UB,3))
```
