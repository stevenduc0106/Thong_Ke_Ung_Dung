```python
# Bài 1:
# Lan rút các quả bóng 5 lần từ một túi 10 quả bóng trong đó có 5 quả bóng màu đỏ và 5 quả bóng màu xanh lá cây thay thế.
# Kiểm tra xem đây có phải là một ví dụ về các thử nghiệm Bernoulli không.

# Giải tay:
# Xác suất rút được bóng đỏ trong lần rút thứ nhất (p):

# Mỗi lần rút có 5 quả bóng đỏ và 5 quả bóng xanh lá cây.
# Vậy xác suất rút được bóng đỏ trong lần đầu tiên là: p = (số quả bóng đỏ) / (tổng số quả bóng) = 5/10 = 0.5.
# Xác suất rút được bóng xanh lá cây trong lần rút thứ nhất (q):

# Xác suất này cũng là 0.5, vì nó là xác suất thất bại (không rút được bóng đỏ) và p + q phải bằng 1.
# Xác suất rút được bóng đỏ trong 5 lần rút liên tiếp:

# Vì các lần rút là độc lập, nên ta có thể tính xác suất này bằng cách nhân các xác suất rút bóng đỏ trong từng lần rút lại với nhau:
# P(rút bóng đỏ trong 5 lần) = p * p * p * p * p = 0.5^5 = 0.03125.
# Xác suất rút được ít nhất một quả bóng đỏ trong 5 lần rút:

# Để tính xác suất này, ta có thể tính xác suất ngược lại, tức là xác suất không rút được bóng đỏ trong cả 5 lần rút và sau đó lấy 1 trừ đi xác suất đó.
# P(rút bóng đỏ trong 5 lần) = (0.5) * (4/9) * (3/8) * (2/7) * (1/6) = 0.00396825396
# Xác suất ít nhất một bóng đỏ = 1 - Xác suất không rút được bóng đỏ = 1 - 0.00396825396 = 0.996
# Vậy, kết quả chính xác là:

# Xác suất rút được bóng đỏ trong lần rút thứ nhất: p = 0.5.
# Xác suất rút được bóng đỏ trong 5 lần rút liên tiếp: 0.00396825396
# Xác suất ít nhất một bóng đỏ trong 5 lần rút:  0.996

# Xác suất rút được bóng đỏ trong lần rút thứ nhất
p = 5/10  # Số quả bóng đỏ chia cho tổng số quả bóng

# Xác suất rút được bóng xanh lá cây trong lần rút thứ nhất
q = 1 - p

# Xác suất rút được bóng đỏ trong 5 lần rút liên tiếp
xacsuatRutDo5Lan = (0.5) * (4/9) * (3/8) * (2/7) * (1/6)

# Xác suất ít nhất một quả bóng đỏ trong 5 lần rút
xacsuatItnhatMotBongDo = 1 - xacsuatRutDo5Lan

print("Xác suất rút được bóng đỏ trong lần rút thứ nhất:", p)
print("Xác suất rút được bóng xanh lá cây trong lần rút thứ nhất:", q)
print("Xác suất rút được bóng đỏ trong 5 lần rút liên tiếp:", xacsuatRutDo5Lan)
print("Xác suất ít nhất một quả bóng đỏ trong 5 lần rút:", xacsuatItnhatMotBongDo)
```

```python
# Bài 2:
# Tỷ lệ người dân tham gia trên mạng lưới giao dịch có sự hiểu biết về luật giao dịch là 90%. Tại một nút giao diện có 5 người giao diện trong phạm vi luật. Gọi X là một số người am hiểu luật nhưng cố gắng tình luật trong 5 người đó.
# a) Lập bảng phân phối của X.
# b) Tính E (3- 5X) và V (3X-2).
# c) Nếu có 30 người trong phạm vi luật, thì trung bình có bao nhiêu người hiểu biết luật nhưng cố gắng trong phạm vi? 
# Có bao nhiêu người trong phạm vi luật làm người không hiểu luật?

import math

# a) Lập bảng phân phối của X.
def binomial_distribution(n, k, p):
    # Tính số cách chọn k phần tử từ n phần tử
    n_choose_k = math.comb(n, k)
    # Tính xác suất P(X = k)
    prob = n_choose_k * (p ** k) * ((1 - p) ** (n - k))
    return prob

n = 5  # Số người giao diện trong phạm vi luật
p = 0.9  # Xác suất thành công (người hiểu luật)

print("a) Bảng phân phối của X:")
print("----------------------------------------")
print("| k (số người am hiểu luật) | P(X = k)  |")
print("----------------------------------------")
for k in range(n + 1):
    prob = binomial_distribution(n, k, p)
    print(f"| {k:2}                         | {prob:.5f}   |")
print("----------------------------------------")

# Tính kỳ vọng của biểu thức 3 - 5X
def tinhKyVong3Tru5X():
    # Kỳ vọng của X đã tính ở trước đó là 4.5
    kyVongX = 4.5
    kyVongKetQua = 3 - 5 * kyVongX
    return kyVongKetQua

# Tính phương sai của biểu thức 3X - 2
def tinhPhuongSai3XTru2():
    # Phương sai của X đã tính ở trước đó là 0.4501
    phuongSaiX = 0.4501
    phuongSaiKetQua = 9 * phuongSaiX  # Phương sai của 3X
    return phuongSaiKetQua

# In kết quả
print("Kỳ vọng của biểu thức 3 - 5X:", tinhKyVong3Tru5X())
print("Phương sai của biểu thức 3X - 2:", tinhPhuongSai3XTru2())



# c) Nếu có 30 người trong phạm vi luật.
soNguoiTrongPhamViLuat = 30

trungBinhHieuLuat = 4.5 * (soNguoiTrongPhamViLuat / 5)

khongHieuLuat = soNguoiTrongPhamViLuat - trungBinhHieuLuat

print("Trung bình số người hiểu luật và cố gắng trong phạm vi 30 người:", trungBinhHieuLuat)
print("Số người trong phạm vi luật không hiểu luật:", khongHieuLuat)

```

```python
# Bài 3:
# Một lô hàng có 8 sản phẩm loai I và 2 sản phẩm loại II. Lấy ngẫu nhiên lần lượt ra 5 sản phẩm theo phương thức hoàn lại. Gọi X là số sản phẩm loại II trong 5 sản phẩm lấy ra.
# a) X có phân phối gì? => X có phân phối nhị thức n = 5 p = 0.2
# b) Tính kỳ vọng và phương sai của X.
# c) Tính số sản phẩm loại II trung bình trong số sản phẩm lấy ra và tính khả năng để xảy ra điều đó.
# d) Nếu lấy lần lượt ra 64 sản phẩm từ lô hàng đó (vẫn lấy theo phương
# thức hoàn lại) thì trung bình lấy được bao nhiêu sản phẩm loại II? Số sản
# phẩm loại II có khả năng xảy ra nhất là bao nhiêu?
# a) Công thức phân phối nhị thức là:

# P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)

# Khi k = 0:
# P(X = 0) = C(5, 0) * (0.2)^0 * (0.8)^5 = 1 * 1 * 0.32768 = 0.32768

# Khi k = 1:
# P(X = 1) = C(5, 1) * (0.2)^1 * (0.8)^4 = 5 * 0.2 * 0.4096 = 0.4096

# Khi k = 2:
# P(X = 2) = C(5, 2) * (0.2)^2 * (0.8)^3 = 10 * 0.04 * 0.512 = 0.2048

# Khi k = 3:
# P(X = 3) = C(5, 3) * (0.2)^3 * (0.8)^2 = 10 * 0.008 * 0.064 = 0.0512

# Khi k = 4:
# P(X = 4) = C(5, 4) * (0.2)^4 * (0.8)^1 = 5 * 0.0016 * 0.8 = 0.0064

# Khi k = 5:
# P(X = 5) = C(5, 5) * (0.2)^5 * (0.8)^0 = 1 * 0.00032 * 1 = 0.00032

import math

# Hàm tính số cách chọn k sản phẩm từ n sản phẩm (tính tổ hợp chập k của n)
def tinhToHopChapK(n, k):
    return math.comb(n, k)

# Hàm tính phân phối xác suất của X
def tinhPhanPhoiXacSuatX(soLanLayMau, xacXuatThanhCong):
    # Số sản phẩm loại I trong lô hàng
    soSanPhamLoaiI = 8
    # Số sản phẩm loại II trong lô hàng
    soSanPhamLoaiII = 2
    
    # Tính xác suất thành công (lấy được sản phẩm loại II)
    p = soSanPhamLoaiII / (soSanPhamLoaiI + soSanPhamLoaiII)
    
    # Khởi tạo danh sách lưu trữ phân phối xác suất
    phanPhoiXacSuat = []
    
    for k in range(soLanLayMau + 1):
        # Công thức phân phối nhị thức
        xacSuat = tinhToHopChapK(soLanLayMau, k) * (p ** k) * ((1 - p) ** (soLanLayMau - k))
        phanPhoiXacSuat.append((k, xacSuat))
    
    return phanPhoiXacSuat

# Các thông số của bài toán
soLanLayMau = 5  # Số lần lấy mẫu (lấy ra 5 sản phẩm)
xacXuatThanhCong = 0.2  # Xác suất thành công trong một lần lấy mẫu (lấy được sản phẩm loại II)

# In kết quả phân phối xác suất của X
print("Phân phối xác suất của X (số sản phẩm loại II trong 5 sản phẩm lấy ra):")
print("-----------------------------------------")
print("| k (số sản phẩm loại II) | P(X = k)  |")
print("-----------------------------------------")
for k, xacSuat in tinhPhanPhoiXacSuatX(soLanLayMau, xacXuatThanhCong):
    print(f"| {k:2}                        | {xacSuat:.5f}   |")
print("-----------------------------------------")

# b) Kỳ vọng (expected value) của X:
# Kỳ vọng được tính bằng công thức: E(X) = n * p

# Trong trường hợp này, n (số sản phẩm được lấy ra) là 5 và p (xác suất lấy sản phẩm loại II trong mỗi lần lấy) 
# là 2/10 = 1/5 (vì có 2 sản phẩm loại II trong tổng cộng 10 sản phẩm).

# E(X) = 5 * (1/5) = 1

# Vậy, kỳ vọng của X là 1.

# Phương sai (variance) của X:
# Phương sai được tính bằng công thức: Var(X) = n * p * (1 - p)

# Trong trường hợp này, n (số sản phẩm được lấy ra) là 5 và p (xác suất lấy sản phẩm loại II trong mỗi lần lấy) 
# là 2/10 = 1/5.

# V(X) = 5 * (1/5) * (1 - 1/5) = 5 * (1/5) * (4/5) = 4/5

# Vậy, phương sai của X là 4/5.
from scipy.stats import binom

# Số lượng sản phẩm loại I trong lô hàng
SoLuongLoaiI = 8
# Số lượng sản phẩm loại II trong lô hàng
SoLuongLoaiII = 2
# Tổng số sản phẩm trong lô hàng
SoLuongTong = SoLuongLoaiI + SoLuongLoaiII
# Số sản phẩm được lấy ra
SoLuongLayRa = 5

# Xác suất lấy sản phẩm loại II trong mỗi lần lấy (p)
XacSuatP = SoLuongLoaiII / SoLuongTong

# Tính kỳ vọng của X
KyVong = SoLuongLayRa * XacSuatP

# Tính phương sai của X
PhuongSai = SoLuongLayRa * XacSuatP * (1 - XacSuatP)

print("Kỳ vọng của X:", KyVong)
print("Phương sai của X:", PhuongSai)
# c) Tính số sản phẩm loại II trung bình trong số sản phẩm lấy ra và tính khả năng để xảy ra điều đó.
# C(2, 1) là cách chọn 1 sản phẩm loại II từ 2 sản phẩm loại II, tức là C(2, 1) = 2! / (1!(2 - 1)!) = 2.
# C(8, 4) là cách chọn 4 sản phẩm loại I từ 8 sản phẩm loại I, tức là C(8, 4) = 8! / (4!(8 - 4)!) = 70.
# Bây giờ chúng ta tính khả năng P(X = 1), tức là khả năng để chọn 1 sản phẩm loại II và 4 sản phẩm loại I 
# từ tổng cộng 5 sản phẩm được lấy ra từ lô hàng:

# P(X = 1) = (C(2, 1) * C(8, 4)) / C(10, 5) = (2 * 70) / (252) = 140 / 252 = 35 / 63.
import math

# Tính tổ hợp chập k của n
def tinhToHopChapK(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

# Số sản phẩm loại I
soSanPhamLoaiI = 8
# Số sản phẩm loại II
soSanPhamLoaiII = 2
# Tổng số sản phẩm
tongSoSanPham = soSanPhamLoaiI + soSanPhamLoaiII
# Số sản phẩm lấy ra
soSanPhamLayRa = 5

# Tính khả năng P(X = 1)
k = 1  # Số sản phẩm loại II
p_X_1 = (tinhToHopChapK(soSanPhamLoaiII, k) * tinhToHopChapK(soSanPhamLoaiI, soSanPhamLayRa - k)) / tinhToHopChapK(tongSoSanPham, soSanPhamLayRa)

print("Khả năng P(X = 1):", p_X_1)
```

```python
# Bài 4:
#a) Xác suất có đúng 5 cuộc điện thoại gọi tới trong 1 giờ là bao nhiêu?

# Chúng ta sử dụng công thức phân phối Poisson:

# P(X = k) = (e^(-λ) * λ^k) / k!

# Ở đây, λ (lambda) là trung bình, trong trường hợp này là 10 cuộc điện thoại trong 1 giờ, và k là số cuộc điện thoại mà chúng ta muốn tính xác suất (k = 5).

# P(X = 5) = (e^(-10) * 10^5) / 5!

# P(X = 5) = (e^(-10) * 100000) / (5 * 4 * 3 * 2 * 1)

# P(X = 5) = (e^(-10) * 100000) / 120

# P(X = 5) ≈ 0.037833

# Vậy, xác suất có đúng 5 cuộc điện thoại gọi tới trong 1 giờ là khoảng 0.037833.

# b) Xác suất có 3 hoặc ít hơn 3 cuộc điện thoại gọi tới trong 1 giờ là bao nhiêu?

# Chúng ta tính tổng xác suất P(X = 0), P(X = 1), và P(X = 2) và sau đó trừ kết quả này từ 1 (tổng xác suất của tất cả các trường hợp):

# P(X ≤ 2) = P(X = 0) + P(X = 1) + P(X = 2)

# P(X = 0) = (e^(-10) * 10^0) / 0! = (e^(-10) * 1) / 1 = e^(-10)

# P(X = 1) = (e^(-10) * 10^1) / 1! = (e^(-10) * 10) / 1 = 10e^(-10)

# P(X = 2) = (e^(-10) * 10^2) / 2! = (e^(-10) * 100) / 2 = 50e^(-10)

# P(X ≤ 2) = e^(-10) + 10e^(-10) + 50e^(-10) = 61e^(-10)

# Vậy, xác suất có 3 hoặc ít hơn 3 cuộc điện thoại gọi tới trong 1 giờ là 61e^(-10).

# c) Xác suất có đúng 15 cuộc điện thoại gọi tới trong 2 giờ là bao nhiêu?

# Ở đây, λ = 20 (10 cuộc/giờ * 2 giờ) và k = 15.

# P(X = 15) = (e^(-20) * 20^15) / 15! ≈ 0.0517 

# P(X = 15) 

# d) Xác suất có đúng 5 cuộc điện thoại gọi tới trong 30 phút là bao nhiêu?

# Ở đây, λ = 5 (10 cuộc/giờ * 0,5 giờ) và k = 5.

# P(X = 5) = (e^(-5) * 5^5) / 5! ≈ 0.1755

import math

# Hàm tính xác suất theo phân phối Poisson
def tinhXacSuatPoisson(k, lambda_):
    return (math.exp(-lambda_) * lambda_**k) / math.factorial(k)

# a) Xác suất có đúng 5 cuộc điện thoại gọi tới trong 1 giờ
lambdaA = 10  # Trung bình 10 cuộc điện thoại trong 1 giờ
soCuocGoiA = 5
xacSuatA = tinhXacSuatPoisson(soCuocGoiA, lambdaA)
print("a) Xác suất có đúng 5 cuộc điện thoại gọi tới trong 1 giờ:", xacSuatA)

# b) Xác suất có 3 hoặc ít hơn 3 cuộc điện thoại gọi tới trong 1 giờ
soCuocGoiB = [0, 1, 2, 3]  # Số cuộc điện thoại từ 0 đến 3
xacSuatB = sum(tinhXacSuatPoisson(k, lambdaA) for k in soCuocGoiB)
print("b) Xác suất có 3 hoặc ít hơn 3 cuộc điện thoại gọi tới trong 1 giờ:", xacSuatB)

# c) Xác suất có đúng 15 cuộc điện thoại gọi tới trong 2 giờ
lambdaC = 20  # Trung bình 20 cuộc điện thoại trong 2 giờ
soCuocGoiC = 15
xacSuatC = tinhXacSuatPoisson(soCuocGoiC, lambdaC)
print("c) Xác suất có đúng 15 cuộc điện thoại gọi tới trong 2 giờ:", xacSuatC)

# d) Xác suất có đúng 5 cuộc điện thoại gọi tới trong 30 phút
lambdaD = 5  # Trung bình 5 cuộc điện thoại trong 30 phút
soCuocGoiD = 5
xacSuatD = tinhXacSuatPoisson(soCuocGoiD, lambdaD)
print("d) Xác suất có đúng 5 cuộc điện thoại gọi tới trong 30 phút:", xacSuatD)
```
