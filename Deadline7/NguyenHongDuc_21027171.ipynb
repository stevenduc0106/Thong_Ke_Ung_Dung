```python
# Bài 2:
# Trong hệ thống tỷ giá hối đoái thả nổi, sự biến đổi của tỷ giá hối đoái thả nổi, sự biến động của tỷ giá hối đoái 
# chịu sự tác động của rất nhiều nhân tố và có thể xem như biến ngẫu nhiên phấn phối chuẩn, giả sử ở một giai đoạn 
# nào đó tỷ giá của USD với VND có trung bình là 15000đ và độ lệch chuẩn là 500đ. Tìm xác suất để trong một ngày nào đó.
# a) Tỷ giá sẽ cao hơn 16000đ
# b) Tỷ giá sẽ thấp hơn 14500đ
# c) Nằm trong khoảng 14500đ đến 16500

# Giải tay:
# μ = 15000
# σ = 500

# a) Chuẩn hóa tỷ giá 16000 VND để đưa nó về phân phối chuẩn:
#    Z = ( X - μ ) / σ = (16000 - 15000) / 500 = 2 

# b) Chuẩn hóa tỷ giá 14500 VND để đưa nó về phân phối chuẩn:
#    Z = ( X - μ ) / σ = (14500 - 15000) / 500 = -1 

# b) Chuẩn hóa tỷ giá 14500đ đến 16500 để đưa nó về phân phối chuẩn:
#    Z1 = ( X - μ ) / σ = (14500 - 15000) / 500 = -1 
#    Z2 = ( X - μ ) / σ = (16500 - 15000) / 500 = 3
import scipy.stats as stats

# Giá trị trung bình và độ lệch chuẩn
mu = 15000
sigma = 500

# a) Tỷ giá sẽ cao hơn 16000đ
Z_a = (16000 - mu) / sigma
probability_a = 1 - stats.norm.cdf(Z_a)
print("a) Xác suất tỷ giá cao hơn 16000 VND:", probability_a)

# b) Tỷ giá sẽ thấp hơn 14500đ
Z_b = (14500 - mu) / sigma
probability_b = stats.norm.cdf(Z_b)
print("b) Xác suất tỷ giá thấp hơn 14500 VND:", probability_b)

# c) Nằm trong khoảng 14500đ đến 16500đ
Z_c1 = (14500 - mu) / sigma
Z_c2 = (16500 - mu) / sigma
probability_c = stats.norm.cdf(Z_c2) - stats.norm.cdf(Z_c1)
print("c) Xác suất tỷ giá nằm trong khoảng 14500 VND đến 16500 VND:", probability_c)
# Vẽ hình phân phối
import matplotlib.pyplot as plt
import numpy as np

# Giá trị Z chuẩn hóa cho 16000 VND, 14500 VND và 16500 VND
Z_a = (16000 - 15000) / 500
Z_b = (14500 - 15000) / 500
Z_c1 = (14500 - 15000) / 500
Z_c2 = (16500 - 15000) / 500

# Tính xác suất cho a)
probability_a = 1 - stats.norm.cdf(Z_a)

# Tính xác suất cho b)
probability_b = stats.norm.cdf(Z_b)

# Tính xác suất cho c)
probability_c = stats.norm.cdf(Z_c2) - stats.norm.cdf(Z_c1)

# Vẽ biểu đồ phân phối xác suất
x = np.linspace(-3, 3, 1000) 
pdf = stats.norm.pdf(x, loc=0, scale=1) 

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Phân phối chuẩn tiêu chuẩn')
plt.fill_between(x, pdf, where=(x < Z_a), alpha=0.2, label='P(Z > Z_a)')
plt.fill_between(x, pdf, where=(x < Z_b), alpha=0.2, label='P(Z < Z_b)')
plt.fill_between(x, pdf, where=((x > Z_c1) & (x < Z_c2)), alpha=0.2, label='P(Z_c1 < Z < Z_c2)')

plt.xlabel('Z (giá trị chuẩn hóa)')
plt.ylabel('Hàm mật độ xác suất')
plt.title('Biểu đồ Phân phối xác suất')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bài 3:
# Chiều cao nam giới khi trưởng thành ở một vùng dân cư là biến ngẫu nhiên phân phối chuẩn với μ=160 cm và σ=6cm . 
# Một thanh niên bị coi là lùn nếu có chiều cao nhỏ hơn 155cm
# a) Tìm tỷ lệ thanh niên lùn ở vùng đó
# b) Tìm xác suất để lấy ngẫu nhiên 4 người thì có ít nhất 1 người không bị lùn.

# Giải tay:
# μ=160
# σ=6cm

# a) Sử dụng phân phối chuẩn, ta tính z-score cho giá trị 155 cm:
# Z = (X - μ) / σ = (155 - 160) / 6 = -5/6 ≈ -0.8333 
# Tra bảng => xác suất ≈ 0.2023
import scipy.stats as stats

# Giá trị Z chuẩn hóa cho 155 cm
Z = (155 - 160) / 6

# Tính xác suất cho a)
probability = stats.norm.cdf(Z)
print("a) Tỷ lệ thanh niên lùn ở vùng đó:", probability)
# b) P(Lùn) = 0.2023 
#    Xác suất 4 người đều lùn: P(4Lùn) = (0.2023)^4 
#    Xác suất ít nhít 1 người không bị lùn = 1 - (0.2023)^4 ≈ 0.9983
import scipy.stats as stats

# Xác suất bị lùn cho một người
p_lun = 0.2023

# Sử dụng công thức đối ngẫu, tính xác suất ít nhất 1 người không bị lùn
p_it_nhat_1_khong_lun = 1 - p_lun**4

print("b)Xác suất ít nhất 1 người không bị lùn:", p_it_nhat_1_khong_lun)

#Vẽ hình phân phối
import matplotlib.pyplot as plt
import numpy as np

# Xác suất bị lùn cho một người
p_lun = 0.2023

# Tạo một danh sách các số người không bị lùn từ 0 đến 4
nguoi_khong_lun = np.arange(0, 5)

# Tính xác suất ít nhất 1 người không bị lùn cho từng trường hợp
p_it_nhat_1_khong_lun = 1 - p_lun**nguoi_khong_lun

# Vẽ biểu đồ
plt.figure(figsize=(8, 6))
plt.plot(nguoi_khong_lun, p_it_nhat_1_khong_lun, marker='o', linestyle='-')
plt.xlabel('Số người không bị lùn')
plt.ylabel('Xác suất ít nhất 1 người không bị lùn')
plt.title('Phân phối xác suất ít nhất 1 người không bị lùn trong 4 người')
plt.grid(True)
plt.show()
```

```python
# Bài 4:
# Chiều dài cá được mô hình hóa bằng phân phối chuẩn N(μ=16 (cm), σ=4 (cm)). Ta cần trả lời các câu hỏi sau:
# a) Xác suất bắt được con cá nhỏ (nhỏ hơn 6 (cm))?
# b) Giả sử, ai bắt được con cá lớn (lớn hơn 20(cm)) sẽ được thưởng. Hỏi xác suất được
# thưởng là bao nhiêu?
# c) Xác suất bắt được con cá vừa (trong khoảng 10-24(cm))?

# Giải tay
# μ= 16
# σ= 4
# a) Xác suất bắt được con cá nhỏ (nhỏ hơn 6 (cm))?
# Ta sử dụng phân phối chuẩn và tính z-score cho giá trị 6 cm:
# Z = (X - μ) / σ = (6 - 16) / 4 = -2.5 => tra bảng 0,0062
# b) Giả sử, ai bắt được con cá lớn (lớn hơn 20(cm)) sẽ được thưởng. Hỏi xác suất được
# thưởng là bao nhiêu?
# Ta tính z-score cho giá trị 20 cm:
# Z = (X - μ) / σ = (20 - 16) / 4 = 1 => tra bảng ≈ 0.8413
# c) Xác suất bắt được con cá vừa (trong khoảng 10-24(cm))?
# Ta tính z-score cho giá trị 10 cm:
# Z1 = (X - μ) / σ = (10 - 16) / 4 = -1.5 => tra bảng 0.0668

# Ta tính z-score cho giá trị 24 cm:
# Z2 = (X - μ) / σ = (24 - 16) / 4 = 2   => tra bảng 0,9772

import scipy.stats as stats

# Giá trị trung bình và độ lệch chuẩn
mu = 16
sigma = 4

# a) Xác suất bắt được con cá nhỏ (nhỏ hơn 6 cm)
# Tính giá trị Z-score cho 6 cm
Z_a = (6 - mu) / sigma
# Tính xác suất
probability_a = stats.norm.cdf(Z_a)
print("a) Xác suất bắt được con cá nhỏ:", probability_a)

# b) Xác suất được thưởng (bắt được con cá lớn hơn 20 cm)
# Tính giá trị Z-score cho 20 cm
Z_b = (20 - mu) / sigma
# Tính xác suất
probability_b = 1 - stats.norm.cdf(Z_b)
print("b) Xác suất được thưởng:", probability_b)

# c) Xác suất bắt được con cá vừa (trong khoảng 10-24 cm)
# Tính giá trị Z-score cho 10 cm và 24 cm
Z_c1 = (10 - mu) / sigma
Z_c2 = (24 - mu) / sigma
# Tính xác suất
probability_c = stats.norm.cdf(Z_c2) - stats.norm.cdf(Z_c1)
print("c) Xác suất bắt được con cá vừa:", probability_c)

# Vẽ hình phân phối
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Giá trị trung bình và độ lệch chuẩn
mu = 16
sigma = 4

# Tạo một dãy giá trị từ 0 đến 30 (độ dài tiêu chuẩn để vẽ phân phối)
x = np.linspace(0, 30, 1000)

# Tính hàm mật độ xác suất (PDF) của phân phối chuẩn tại mỗi giá trị x
pdf = stats.norm.pdf(x, mu, sigma)

# Vẽ đường cong phân phối
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Phân phối chuẩn N(16, 4)', color='blue')
plt.xlabel('Chiều dài cá (cm)')
plt.ylabel('Hàm mật độ xác suất')
plt.title('Phân phối chiều dài của cá')
plt.fill_between(x, pdf, where=(x < 6), color='red', alpha=0.3, label='Con cá nhỏ (6 cm)')
plt.fill_between(x, pdf, where=(x >= 6) & (x <= 20), color='green', alpha=0.3, label='Con cá vừa (6-20 cm)')
plt.fill_between(x, pdf, where=(x > 20), color='purple', alpha=0.3, label='Con cá lớn (20 cm)')
plt.legend()
plt.grid(True)
plt.show()
```
