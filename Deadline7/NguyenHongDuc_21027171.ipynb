```python
# Bi 2:
# Trong h thng t gi hi oi th ni, s bin i ca t gi hi oi th ni, s bin ng ca t gi hi oi 
# chu s tc ng ca rt nhiu nhn t v c th xem nh bin ngu nhin phn phi chun, gi s  mt giai on 
# no  t gi ca USD vi VND c trung bnh l 15000 v  lch chun l 500. Tm xc sut  trong mt ngy no .
# a) T gi s cao hn 16000
# b) T gi s thp hn 14500
# c) Nm trong khong 14500 n 16500

# Gii tay:
#  = 15000
#  = 500

# a) Chun ha t gi 16000 VND  a n v phn phi chun:
#    Z = ( X -  ) /  = (16000 - 15000) / 500 = 2 

# b) Chun ha t gi 14500 VND  a n v phn phi chun:
#    Z = ( X -  ) /  = (14500 - 15000) / 500 = -1 

# b) Chun ha t gi 14500 n 16500  a n v phn phi chun:
#    Z1 = ( X -  ) /  = (14500 - 15000) / 500 = -1 
#    Z2 = ( X -  ) /  = (16500 - 15000) / 500 = 3
import scipy.stats as stats

# Gi tr trung bnh v  lch chun
mu = 15000
sigma = 500

# a) T gi s cao hn 16000
Z_a = (16000 - mu) / sigma
probability_a = 1 - stats.norm.cdf(Z_a)
print("a) Xc sut t gi cao hn 16000 VND:", probability_a)

# b) T gi s thp hn 14500
Z_b = (14500 - mu) / sigma
probability_b = stats.norm.cdf(Z_b)
print("b) Xc sut t gi thp hn 14500 VND:", probability_b)

# c) Nm trong khong 14500 n 16500
Z_c1 = (14500 - mu) / sigma
Z_c2 = (16500 - mu) / sigma
probability_c = stats.norm.cdf(Z_c2) - stats.norm.cdf(Z_c1)
print("c) Xc sut t gi nm trong khong 14500 VND n 16500 VND:", probability_c)
# V hnh phn phi
import matplotlib.pyplot as plt
import numpy as np

# Gi tr Z chun ha cho 16000 VND, 14500 VND v 16500 VND
Z_a = (16000 - 15000) / 500
Z_b = (14500 - 15000) / 500
Z_c1 = (14500 - 15000) / 500
Z_c2 = (16500 - 15000) / 500

# Tnh xc sut cho a)
probability_a = 1 - stats.norm.cdf(Z_a)

# Tnh xc sut cho b)
probability_b = stats.norm.cdf(Z_b)

# Tnh xc sut cho c)
probability_c = stats.norm.cdf(Z_c2) - stats.norm.cdf(Z_c1)

# V biu  phn phi xc sut
x = np.linspace(-3, 3, 1000) 
pdf = stats.norm.pdf(x, loc=0, scale=1) 

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Phn phi chun tiu chun')
plt.fill_between(x, pdf, where=(x < Z_a), alpha=0.2, label='P(Z > Z_a)')
plt.fill_between(x, pdf, where=(x < Z_b), alpha=0.2, label='P(Z < Z_b)')
plt.fill_between(x, pdf, where=((x > Z_c1) & (x < Z_c2)), alpha=0.2, label='P(Z_c1 < Z < Z_c2)')

plt.xlabel('Z (gi tr chun ha)')
plt.ylabel('Hm mt  xc sut')
plt.title('Biu  Phn phi xc sut')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bi 3:
# Chiu cao nam gii khi trng thnh  mt vng dn c l bin ngu nhin phn phi chun vi =160 cm v =6cm . 
# Mt thanh nin b coi l ln nu c chiu cao nh hn 155cm
# a) Tm t l thanh nin ln  vng 
# b) Tm xc sut  ly ngu nhin 4 ngi th c t nht 1 ngi khng b ln.

# Gii tay:
# =160
# =6cm

# a) S dng phn phi chun, ta tnh z-score cho gi tr 155 cm:
# Z = (X - ) /  = (155 - 160) / 6 = -5/6  -0.8333 
# Tra bng => xc sut  0.2023
import scipy.stats as stats

# Gi tr Z chun ha cho 155 cm
Z = (155 - 160) / 6

# Tnh xc sut cho a)
probability = stats.norm.cdf(Z)
print("a) T l thanh nin ln  vng :", probability)
# b) P(Ln) = 0.2023 
#    Xc sut 4 ngi u ln: P(4Ln) = (0.2023)^4 
#    Xc sut t nht 1 ngi khng b ln = 1 - (0.2023)^4  0.9983
import scipy.stats as stats

# Xc sut b ln cho mt ngi
p_lun = 0.2023

# S dng cng thc i ngu, tnh xc sut t nht 1 ngi khng b ln
p_it_nhat_1_khong_lun = 1 - p_lun**4

print("b)Xc sut t nht 1 ngi khng b ln:", p_it_nhat_1_khong_lun)

#V hnh phn phi
import matplotlib.pyplot as plt
import numpy as np

# Xc sut b ln cho mt ngi
p_lun = 0.2023

# To mt danh sch cc s ngi khng b ln t 0 n 4
nguoi_khong_lun = np.arange(0, 5)

# Tnh xc sut t nht 1 ngi khng b ln cho tng trng hp
p_it_nhat_1_khong_lun = 1 - p_lun**nguoi_khong_lun

# V biu 
plt.figure(figsize=(8, 6))
plt.plot(nguoi_khong_lun, p_it_nhat_1_khong_lun, marker='o', linestyle='-')
plt.xlabel('S ngi khng b ln')
plt.ylabel('Xc sut t nht 1 ngi khng b ln')
plt.title('Phn phi xc sut t nht 1 ngi khng b ln trong 4 ngi')
plt.grid(True)
plt.show()
```

```python
# Bi 4:
# Chiu di c c m hnh ha bng phn phi chun N(=16 (cm), =4 (cm)). Ta cn tr li cc cu hi sau:
# a) Xc sut bt c con c nh (nh hn 6 (cm))?
# b) Gi s, ai bt c con c ln (ln hn 20(cm)) s c thng. Hi xc sut c
# thng l bao nhiu?
# c) Xc sut bt c con c va (trong khong 10-24(cm))?

# Gii tay
# = 16
# = 4
# a) Xc sut bt c con c nh (nh hn 6 (cm))?
# Ta s dng phn phi chun v tnh z-score cho gi tr 6 cm:
# Z = (X - ) /  = (6 - 16) / 4 = -2.5 => tra bng 0,0062
# b) Gi s, ai bt c con c ln (ln hn 20(cm)) s c thng. Hi xc sut c
# thng l bao nhiu?
# Ta tnh z-score cho gi tr 20 cm:
# Z = (X - ) /  = (20 - 16) / 4 = 1 => tra bng  0.8413
# c) Xc sut bt c con c va (trong khong 10-24(cm))?
# Ta tnh z-score cho gi tr 10 cm:
# Z1 = (X - ) /  = (10 - 16) / 4 = -1.5 => tra bng 0.0668

# Ta tnh z-score cho gi tr 24 cm:
# Z2 = (X - ) /  = (24 - 16) / 4 = 2   => tra bng 0,9772

import scipy.stats as stats

# Gi tr trung bnh v  lch chun
mu = 16
sigma = 4

# a) Xc sut bt c con c nh (nh hn 6 cm)
# Tnh gi tr Z-score cho 6 cm
Z_a = (6 - mu) / sigma
# Tnh xc sut
probability_a = stats.norm.cdf(Z_a)
print("a) Xc sut bt c con c nh:", probability_a)

# b) Xc sut c thng (bt c con c ln hn 20 cm)
# Tnh gi tr Z-score cho 20 cm
Z_b = (20 - mu) / sigma
# Tnh xc sut
probability_b = 1 - stats.norm.cdf(Z_b)
print("b) Xc sut c thng:", probability_b)

# c) Xc sut bt c con c va (trong khong 10-24 cm)
# Tnh gi tr Z-score cho 10 cm v 24 cm
Z_c1 = (10 - mu) / sigma
Z_c2 = (24 - mu) / sigma
# Tnh xc sut
probability_c = stats.norm.cdf(Z_c2) - stats.norm.cdf(Z_c1)
print("c) Xc sut bt c con c va:", probability_c)

# V hnh phn phi
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Gi tr trung bnh v  lch chun
mu = 16
sigma = 4

# To mt dy gi tr t 0 n 30 ( di tiu chun  v phn phi)
x = np.linspace(0, 30, 1000)

# Tnh hm mt  xc sut (PDF) ca phn phi chun ti mi gi tr x
pdf = stats.norm.pdf(x, mu, sigma)

# V ng cong phn phi
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Phn phi chun N(16, 4)', color='blue')
plt.xlabel('Chiu di c (cm)')
plt.ylabel('Hm mt  xc sut')
plt.title('Phn phi chiu di ca c')
plt.fill_between(x, pdf, where=(x < 6), color='red', alpha=0.3, label='Con c nh (6 cm)')
plt.fill_between(x, pdf, where=(x >= 6) & (x <= 20), color='green', alpha=0.3, label='Con c va (6-20 cm)')
plt.fill_between(x, pdf, where=(x > 20), color='purple', alpha=0.3, label='Con c ln (20 cm)')
plt.legend()
plt.grid(True)
plt.show()
```
